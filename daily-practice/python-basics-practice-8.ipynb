## Part 1: 리스트 실전 문제

### 문제 1-1: 양수만 필터링
리스트에서 양수만 골라서 반환

````python
def filter_positive(nums: list) -> list:
    # 여기에 코드 작성
    result = []
    for num in nums:
        if num > 0:
            result.append(num)
    return result

# 테스트
print(filter_positive([1, -2, 3, -4, 5]))  # [1, 3, 5]
print(filter_positive([-1, -2, -3]))       # []
````

### 문제 1-2: 리스트 요소의 곱
리스트의 모든 숫자를 곱한 결과 반환

````python
def multiply_list(nums: list) -> int:
    # 힌트: 초기값을 1로 설정
    result = 1
    for num in nums:
        result *= num
    return result
# 테스트
print(multiply_list([2, 3, 4]))  # 24
print(multiply_list([1, 5, 10])) # 50
````

### 문제 1-3: 특정 값 개수 세기
리스트에서 target 값이 몇 번 나오는지 반환

````python
def count_value(nums: list, target: int) -> int:
    # 여기에 코드 작성
    result = 0
    for num in nums:
        if num == target:
            result += 1
    return result 

# 테스트
print(count_value([1, 2, 3, 2, 4, 2], 2))  # 3
print(count_value([5, 5, 5], 5))           # 3
````

## Part 2: 문자열 실전 문제

### 문제 2-1: 공백 개수 세기
문자열에서 공백(' ')이 몇 개인지 반환

````python
def count_spaces(s: str) -> int:
    # 여기에 코드 작성
    result = 0
    # str = s.split()
    # result = len(str) - 1
    # return result 
    for char in s:
        if char == ' ':
            result += 1
    return result 

# 테스트
print(count_spaces("hello world"))      # 1
print(count_spaces("I love Python"))   # 2
````

### 문제 2-2: 숫자만 추출
문자열에서 숫자만 추출해서 새 문자열로 반환

````python
def extract_digits(s: str) -> str:
    # 힌트: char.isdigit() 사용
    result = ""
    for char in s:
        if char.isdigit():
            result += char

    return result 
# 테스트
print(extract_digits("abc123xyz"))   # "123"
print(extract_digits("hello2024"))   # "2024"
````

### 문제 2-3: 단어 길이 리스트
문장의 각 단어 길이를 리스트로 반환

````python
def word_lengths(sentence: str) -> list:
    # 힌트: split() 후 각 단어의 len() 구하기
    result = []
    split_sentence = sentence.split()

    for char in split_sentence:
        result.append(len(char))
    return result

# 테스트
print(word_lengths("hello world"))     # [5, 5]
print(word_lengths("I love Python"))  # [1, 4, 6]
````

## Part 3: 딕셔너리 실전 문제

### 문제 3-1: 성적 등급 매기기
점수에 따라 등급(A, B, C, D, F) 반환

````python
def get_grade(score: int) -> str:
    # 90 이상: A, 80~89: B, 70~79: C, 60~69: D, 60 미만: F
    result = ""
    if score >= 90:
        return "A"
    elif score < 90 and score >= 80:
        return "B"
    elif score < 80 and score >= 70:
        return "C"
    elif score < 70 and score >= 60:
        return "D"
    else:
        return "F"

# 테스트
print(get_grade(95))  # "A"
print(get_grade(82))  # "B"
print(get_grade(55))  # "F"
````

### 문제 3-2: 알파벳 빈도수
문자열에서 알파벳만 세서 딕셔너리로 반환 (대소문자 구분 X)

````python
def letter_frequency(s: str) -> dict:
    # 힌트: char.isalpha(), lower() 사용
    result = {}

    # for char in s:
    #     if char in result:
    #         result[char] += 1 # 딕셔너리 키값에 값집어넣을때 인덱스처럼 대괄호 사용
    #     else:
    #         result[char] = 1

    for char in s:
        if char.isalpha():
            char = char.lower()
            if char in result:
                result[char] += 1
            else: 
                result[char] = 1
                
    return result 

# 테스트
print(letter_frequency("Hello World"))  # {'h': 1, 'e': 1, 'l': 3, 'o': 2, 'w': 1, 'r': 1, 'd': 1}
print(letter_frequency("AAA bbb"))      # {'a': 3, 'b': 3}
````

### 문제 3-3: 리스트를 딕셔너리로
리스트의 값을 키로, 인덱스를 값으로 하는 딕셔너리 만들기

````python
def list_to_dict_index(items: list) -> dict:
    # 힌트: enumerate 사용
    result = {}
    for index, key in enumerate(items):
        result[key] = index
    return result 

# 테스트
print(list_to_dict_index(['a', 'b', 'c']))  # {'a': 0, 'b': 1, 'c': 2}
print(list_to_dict_index(['x', 'y']))       # {'x': 0, 'y': 1}
````

## Part 4: enumerate 활용

### 문제 4-1: 짝수 인덱스 값만 추출
인덱스가 짝수(0, 2, 4...)인 값들만 반환

````python
def even_indices(nums: list) -> list:
    # 여기에 코드 작성
    result = []
    for index, key in enumerate(nums):
        if index % 2 == 0:
            result.append(key)
    return result 

# 테스트
print(even_indices([10, 20, 30, 40, 50]))  # [10, 30, 50]
print(even_indices([1, 2, 3, 4]))          # [1, 3]
````

### 문제 4-2: 인덱스와 값 곱하기
각 원소에 인덱스를 곱한 리스트 반환

````python
def multiply_by_index(nums: list) -> list:
    # 예: [5, 6, 7] -> [0, 6, 14] (인덱스 0*5, 1*6, 2*7)
    result = []
    for index, key in enumerate(nums):
        result.append(index * key)
    return result

# 테스트
print(multiply_by_index([5, 6, 7]))  # [0, 6, 14]
print(multiply_by_index([2, 3, 4]))  # [0, 3, 8]
````
