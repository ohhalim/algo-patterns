## Part 1: while 반복문

### 문제 1-1: 1부터 n까지 합
1부터 n까지의 합을 반환 (while문 사용)

````python
def sum_to_n(n: int) -> int:
    # while문으로 풀기
    sum_n = 0
    # 내가 생각했던 방법
    # while i in n:
    #     sum_n += i
    # return sum_n 

    i = 1 # 카운터 초기값
    while i <= n: # 조건 (언제까지 ?)
        sum_n += i
        i += 1 # 증가
    return sum_n


# 테스트
print(sum_to_n(5))   # 15 (1+2+3+4+5)
print(sum_to_n(10))  # 55
````

### 문제 1-2: 자릿수 합
정수의 각 자릿수 합 반환 (예: 123 → 1+2+3 = 6)

````python
def digit_sum(n: int) -> int:
    # 힌트: n % 10 = 마지막 자리, n // 10 = 나머지 자리들
    sum_n = 0
    while n > 0:
        sum_n += n % 10
        n = n // 10

    return sum_n
# 테스트
print(digit_sum(123))   # 6
print(digit_sum(9999))  # 36
````

## Part 2: enumerate 활용

### 문제 2-1: 인덱스와 값 같이 출력
리스트의 인덱스와 값을 튜플로 반환

````python
def index_value_pairs(lst: list) -> list:
    # enumerate 사용!
    result = []
    for index, value in enumerate(lst):
        result.append((index, value))
    return result

# 테스트
print(index_value_pairs(['a', 'b', 'c']))  # [(0, 'a'), (1, 'b'), (2, 'c')]
````

### 문제 2-2: 특정 값의 인덱스 찾기
target이 있는 모든 인덱스를 리스트로 반환

````python
def find_all_indices(lst, target):
    result = []
    for i, val in enumerate(lst):  # enumerate 키, 값을 순회하는것
        if val == target: # 값과 타겟이 같다면 
            result.append(i) # 해당 인덱스를 리드트에 넣어라
    return result

# 테스트
print(find_all_indices([1, 2, 3, 2, 4, 2], 2))  # [1, 3, 5]
print(find_all_indices([1, 1, 1], 1))           # [0, 1, 2]
````

## Part 3: 문자열 처리

### 문제 3-1: 모음 개수 세기
문자열에서 모음(a, e, i, o, u) 개수 반환

````python
def count_vowels(s):
    count = 0
    for char in s:
        if char in 'aeiou'
            count += 1
    return count 

# 테스트
print(count_vowels("hello"))      # 2
print(count_vowels("aeiou"))      # 5
print(count_vowels("rhythm"))     # 0
````

### 문제 3-2: 단어 뒤집기
문장의 각 단어를 뒤집기 (단어 순서는 유지)

````python
def reverse_words(s):
    # 힌트: split(), 슬라이싱 [::-1], join()
    words =s.split()
    result = []
    for word in words:
        result.append(word[::-1])
    return ' '.join(result)
    # split() 문자열을 리스트로 나눈다
    # join() 리스트를 문자열로 바꿈

# 테스트
print(reverse_words("hello world"))  # "olleh dlrow"
print(reverse_words("I love python"))  # "I evol nohtyp"
````

### 문제 3-3: 팰린드롬 체크
앞뒤로 읽어도 같은지 확인 (True/False)

````python
def is_palindrome(s):
    return s == s[::-1]

# 테스트
print(is_palindrome("racecar"))  # True
print(is_palindrome("hello"))    # False
print(is_palindrome("a"))        # True
````

## Part 4: 리스트 심화

### 문제 4-1: 두 리스트 합치기
두 리스트를 번갈아가며 합치기

````python
def interleave(list1, list2):
    # 힌트: 둘 다 같은 길이라고 가정
    result = []
    for i in range(len(list1)): # range 함수는 슬라이싱이랑 비슷하다 range(초기값, 마지막값, 증감값)
        result.append(list1[i])
        result.append(list2[i])
    return result

# 테스트
print(interleave([1, 2, 3], ['a', 'b', 'c']))  # [1, 'a', 2, 'b', 3, 'c']
````

### 문제 4-2: 홀수 인덱스 값만 추출
인덱스가 홀수인 값들만 반환

````python
def odd_indices(lst):
    result = []
    for i, val in enumerate(lst):
        if i % 2 == 1:
            result.append(val)
    return result

# 테스트
print(odd_indices([10, 20, 30, 40, 50]))  # [20, 40]
````

### 문제 4-3: 연속 중복 제거
연속으로 같은 값이 있으면 하나만 남기기

````python
def remove_consecutive_duplicates(lst):
    if not lst:
        return []
    result = [lest[0]]
    for i in range(1, len(lst)):
        if lst[i] != lst[i-1]:
            result.append(lst[i])
    return result

# 테스트
print(remove_consecutive_duplicates([1, 1, 2, 2, 2, 3, 1]))  # [1, 2, 3, 1]
````

## Part 5: 딕셔너리 심화

### 문제 5-1: 두 리스트로 딕셔너리 만들기
keys와 values로 딕셔너리 생성

````python
def lists_to_dict(keys, values):
    result = {}
    for i in range(len(keys)):
        result[keys[i]] = values[i]
    return result

# 테스트
print(lists_to_dict(['a', 'b', 'c'], [1, 2, 3]))  # {'a': 1, 'b': 2, 'c': 3}
````

### 문제 5-2: 딕셔너리 키/값 뒤집기
키와 값을 서로 바꾸기

````python
def invert_dict(d):
    result = {}
    for key, value in d.items(): # 딕셔너리의 키,값을  순회하는 메서드. emunerate의 인덱스와 값을 순회하는것과 비슷
        result[value] = key
    return result
# 테스트
print(invert_dict({'a': 1, 'b': 2}))  # {1: 'a', 2: 'b'}
````
