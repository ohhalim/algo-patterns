## Part 1: 조건문 (if/else)

### 문제 1-1: 짝수/홀수 판별
숫자가 짝수면 "even", 홀수면 "odd" 반환

````python
def check_even_odd(n):
    # 여기에 코드 작성
    if n % 2 == 0:
        return "even"
    else:
        return "add"

# 테스트
print(check_even_odd(4))  # "even"
print(check_even_odd(7))  # "odd"
````

### 문제 1-2: 양수/음수/0 판별
양수면 "positive", 음수면 "negative", 0이면 "zero" 반환

````python
def check_sign(n):
    # 여기에 코드 작성
    if n > 0:
        return "positive"
    elif n < 0:
        return "negative"
    else:
        return "zero"

# 테스트
print(check_sign(5))   # "positive"
print(check_sign(-3))  # "negative"
print(check_sign(0))   # "zero"
````

### 문제 1-3: 최댓값 찾기
두 숫자 중 큰 값 반환

````python
def find_max(a, b):
    # 여기에 코드 작성
    if a > b:
        return a 
    else:
        return b

# 테스트
print(find_max(3, 7))  # 7
print(find_max(10, 2)) # 10
````

## Part 2: 반복문 (for)

### 문제 2-1: 리스트 합계
리스트의 모든 숫자 합 반환

````python
def sum_list(nums):
    # 여기에 코드 작성
    sum_list = int() # 이거 잘안씀 0으로 많이 씀
    for i in nums:
        sum_list += i 
    return sum_list
# 테스트
print(sum_list([1, 2, 3, 4, 5]))  # 15
print(sum_list([10, 20, 30]))     # 60
````

### 문제 2-2: 리스트에서 최댓값 찾기
리스트에서 가장 큰 값 반환 (max() 사용 금지!)

````python
def find_max_in_list(nums):
    # 여기에 코드 작성
    # a = sorted(nums)
    # return a[::-1]
    max_num = 0 # nums[0] 이거로 처리하는게 더 낫다. 0보다 더 작은 수로만 되어있을수있으니까
    for i in nums:
        if i > max_num:
            max_num = i
    return max_num


# 테스트
print(find_max_in_list([3, 1, 4, 1, 5, 9]))  # 9
print(find_max_in_list([10, 2, 8]))          # 10
````

### 문제 2-3: 짝수만 필터링
리스트에서 짝수만 담은 새 리스트 반환

````python
def filter_even(nums):
    # 여기에 코드 작성
    even_nums = []
    for i in nums:
        if i % 2 == 0:
            even_nums.append(i)
    return even_nums 
    # 기분개좋다 완벽하게 풀었어 스트레스 풀린다
    # 이래서 코딩이 좋아 해결했어 이게 내 결과물이야 너무기분좋아 작지먼 내가 생각해서 문제를 해결했어 
# 테스트
print(filter_even([1, 2, 3, 4, 5, 6]))  # [2, 4, 6]
print(filter_even([7, 8, 9, 10]))       # [8, 10]
````

### 문제 2-4: 문자열 뒤집기
문자열을 거꾸로 반환 (슬라이싱 금지, for문 사용)

````python
def reverse_string(s):
    # 여기에 코드 작성
    reverse_str = "" # int는 0으로 문자열은 ""이거로 함
    # for str in s[::-1]:
    #     reverse_str += str
    # return reverse_str
    for str in s:
        reverse_str = str + reverse_str # 초기화 문자열 앞에 하나씩 붙이면됨
    return reverse_str
print(reverse_string("hello"))  # "olleh"
print(reverse_string("python")) # "nohtyp"
````

## Part 3: 딕셔너리 (dict)

### 문제 3-1: 문자 개수 세기
문자열에서 각 문자가 몇 번 나오는지 딕셔너리로 반환

````python
def count_chars(s):
    # 여기에 코드 작성
    count = {}
    for char in s:
        if char in count: 
            count[char] += 1
        else:
            count[char] = 1
    return count

# 테스트
print(count_chars("hello"))  # {'h': 1, 'e': 1, 'l': 2, 'o': 1}
print(count_chars("aab"))    # {'a': 2, 'b': 1}
````

### 문제 3-2: 딕셔너리 값 합계
딕셔너리의 모든 값(value) 합 반환

````python
def sum_values(d):
    # 이거 맞는데 더 맞게 함수명을 바꾸기
    # sum_dict_velue = 0
    # for i in d:
    #     sum_dict_velue += d[i]      
    # return sum_dict_velue 

    # 더 명확하게
    total = 0 
    for key in d:
        total += d[key]
    return total
# 테스트
print(sum_values({'a': 1, 'b': 2, 'c': 3}))  # 6
print(sum_values({'x': 10, 'y': 20}))        # 30
````

## Part 4: Set

### 문제 4-1: 중복 제거
리스트에서 중복을 제거한 리스트 반환

````python
def remove_duplicates(nums) -> list[int]:
    # 여기에 코드 작성
    # 리스트로 반환해야해서 리스트로 더 씌워줘야해 return set(nums)
    return list(set(nums))


# 테스트
print(sorted(remove_duplicates([1, 2, 2, 3, 3, 3])))  # [1, 2, 3]
print(sorted(remove_duplicates([1, 1, 1])))           # [1]
````

## Part 5: 종합 문제

### 문제 5-1: 가장 많이 나온 문자
문자열에서 가장 많이 나온 문자 반환

````python
def most_frequent_char(s) -> str():
    # 여기에 코드 작성
    # dict_s 개수 센 키, 벨류 딕셔너리임
    dict_s = {}
    for char in s:
        if char in dict_s: 
            dict_s[char] += 1
        else:
            dict_s[char] = 1
    # 이건 알파벳순을 정렬한거임 틀린거 
    # a = sorted(dict_s) 
    # return a[::-1]

    # 초기값 
    max_char = s[0]
    # items 키, 값 같이 순회 한다는 뜻

    # dict_s을 순회한다 키, 값을 둘다
    for char, count in dict_s.items():
        # 만약 값이 딕셔너리 초기값보다 크다면(이해를 위해서 초기값이라 말한거고 순회하면서 계속 비교하며 높은 값을 가지고 있는 키가 바뀜)
        if count > dict_s[max_char]:
            # 그 키는 바뀐다 
            max_char = char 

    return max_char

# 테스트
print(most_frequent_char("aabbbcc"))  # 'b'
print(most_frequent_char("hello"))    # 'l'
````

### 문제 5-2: 리스트 뒤집기 (in-place)
새 리스트 만들지 않고 원본 리스트를 뒤집기

````python
def reverse_list_inplace(nums) -> list[int]:
    # 여기에 코드 작성 (nums 자체를 수정)

    # reversed()는 새 객체를 만들어서 원본을 안바꿔
    # return reversed(nums)
    left = 0
    right= len(nums) - 1

    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1


# 테스트
test1 = [1, 2, 3, 4, 5]
reverse_list_inplace(test1)
print(test1)  # [5, 4, 3, 2, 1]
````
