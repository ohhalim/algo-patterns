### 문제 1: 팩토리얼
n! = n * (n-1) * (n-2) * ... * 1 계산 (while문 사용)

````python
def factorial(n: int) -> int:
    # while문으로 풀기
    sum_n = 1
    i = 1
    while i <= n: # i가 n보다 작거나 같을때
        sum_n *= i
        i = i+1
    return sum_n


# 테스트
print(factorial(5))   # 120 (5*4*3*2*1)
print(factorial(3))   # 6 (3*2*1)
print(factorial(0))   # 1 (0! = 1)
````

### 문제 2: 숫자 뒤집기
123 → 321 로 뒤집기 (문자열 변환 없이!)

````python
def reverse_number(n: int) -> int:
    # 힌트: % 10과 // 10 사용, 결과에 10을 곱하면서 자릿수 늘리기
    result = 0 
    while n > 0:
        result = result * 10 + (n% 10)
        n = n//10
    return result
# 테스트
print(reverse_number(123))   # 321
print(reverse_number(1000))  # 1
print(reverse_number(5))     # 5
````
