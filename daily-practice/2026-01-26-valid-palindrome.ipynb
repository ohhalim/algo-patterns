## 문제: 유효한 팰린드롬 (Valid Palindrome)

### 난이도: Easy

---

### 문제 설명

문자열 `s`가 주어집니다.  
**알파벳과 숫자만 고려**하고, **대소문자를 무시**했을 때 앞에서 읽으나 뒤에서 읽으나 같으면 `True`, 아니면 `False`를 반환하세요.

---

### 예제

**예제 1:**
```
입력: s = "A man, a plan, a canal: Panama"
출력: True
설명: 알파벳만 보면 "amanaplanacanalpanama" -> 뒤집어도 같음!
```

**예제 2:**
```
입력: s = "race a car"
출력: False
설명: "raceacar" -> 뒤집으면 "racaecar" -> 다름
```

**예제 3:**
```
입력: s = " "
출력: True
설명: 알파벳/숫자가 없으면 빈 문자열 -> 팰린드롬
```

---

### 제약 조건

- `1 <= s.length <= 2 * 10^5`
- `s`는 ASCII 문자로만 구성됨

---

### 힌트

1. 문자가 알파벳/숫자인지 확인: `char.isalnum()`
2. 소문자로 변환: `char.lower()`
3. 문자열 뒤집기: `s[::-1]`
4. 투 포인터로도 풀 수 있어요!

````python
def is_palindrome(s: str) -> bool:
    """
    문자열이 팰린드롬인지 확인합니다.
    알파벳과 숫자만 고려하고, 대소문자는 무시합니다.
    
    Args:
        s: 입력 문자열
    
    Returns:
        팰린드롬이면 True, 아니면 False
    """
    # 여기에 코드를 작성하세요
    cleand = ''.join(c.lower() for c in s if c.isalnum())
    return cleand    == cleand  [::-1]
    
````

---

## 테스트 케이스

````python
# 테스트 실행
def test_is_palindrome():
    # 테스트 케이스 1
    result1 = is_palindrome("A man, a plan, a canal: Panama")
    assert result1 == True, f"테스트 1 실패: {result1}"
    print("테스트 1 통과!")
    
    # 테스트 케이스 2
    result2 = is_palindrome("race a car")
    assert result2 == False, f"테스트 2 실패: {result2}"
    print("테스트 2 통과!")
    
    # 테스트 케이스 3
    result3 = is_palindrome(" ")
    assert result3 == True, f"테스트 3 실패: {result3}"
    print("테스트 3 통과!")
    
    # 테스트 케이스 4
    result4 = is_palindrome("0P")
    assert result4 == False, f"테스트 4 실패: {result4}"
    print("테스트 4 통과!")
    
    print("\n모든 테스트 통과!")

test_is_palindrome()
````

```python
def is_palindrome(s):
    # 알파벳/숫자만 추출하고 소문자로 변환
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    # 원본과 뒤집은 것 비교
    return cleaned == cleaned[::-1]
```

### 풀이 2: 투 포인터 (공간 효율적)
```python
def is_palindrome(s):
    left, right = 0, len(s) - 1
    
    while left < right:
        # 왼쪽에서 알파벳/숫자 찾기
        while left < right and not s[left].isalnum():
            left += 1
        # 오른쪽에서 알파벳/숫자 찾기
        while left < right and not s[right].isalnum():
            right -= 1
        # 비교
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    
    return True
```

### 핵심 포인트
- `isalnum()`: 알파벳 또는 숫자인지 확인
- `lower()`: 소문자로 변환
- `[::-1]`: 문자열 뒤집기
- 투 포인터는 추가 공간 없이 O(1) 공간복잡도
