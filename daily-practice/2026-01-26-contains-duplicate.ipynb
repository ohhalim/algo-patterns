## 문제: 중복 확인 (Contains Duplicate)

### 난이도: Easy

---

### 문제 설명

정수 배열 `nums`가 주어집니다.  
배열에 **중복된 값이 있으면** `True`, **없으면** `False`를 반환하세요.

---

### 예제

**예제 1:**
```
입력: nums = [1, 2, 3, 1]
출력: True
설명: 1이 두 번 나옴
```

**예제 2:**
```
입력: nums = [1, 2, 3, 4]
출력: False
설명: 모든 값이 고유함
```

**예제 3:**
```
입력: nums = [1, 1, 1, 3, 3, 4, 3, 2, 4, 2]
출력: True
```

---

### 제약 조건

- `1 <= nums.length <= 10^5`
- `-10^9 <= nums[i] <= 10^9`

---

### 힌트

1. 중복이 없으면 원본 길이 == 고유한 값 개수
2. Python의 `set()`은 중복을 제거해줌!
3. `len(nums)` vs `len(set(nums))` 비교해보면?

````python
# def contains_duplicate(nums: list[int]) -> bool:
#     """
#     배열에 중복된 값이 있는지 확인합니다.
    
#     Args:
#         nums: 정수 배열
    
#     Returns:
#         중복이 있으면 True, 없으면 False
#     """
#     # 여기에 코드를 작성하세요

#     pass


def sulution(nums: list[int]) -> bool:
# 배열은 list다 
    seen = set()
    # nums 안에있는 각각의 값에 대해서 
    # nums을 하나씩 돌면서
    for i in nums:
        # 만약 i값이 seen 안에 있다면 
        if i in seen:
            return True
        else:
            seen.add(i)
    # 모든 숫자를 다확인한후에 실행되어이야해서 for문에 있어야함
    return False
````

---

## 테스트 케이스

````python
# 테스트 실행
def test_contains_duplicate():
    # 테스트 케이스 1
    result1 = sulution([1, 2, 3, 1])
    assert result1 == True, f"테스트 1 실패: {result1}"
    print("테스트 1 통과!")
    
    # 테스트 케이스 2
    result2 = sulution([1, 2, 3, 4])
    assert result2 == False, f"테스트 2 실패: {result2}"
    print("테스트 2 통과!")
    
    # 테스트 케이스 3
    result3 = sulution([1, 1, 1, 3, 3, 4, 3, 2, 4, 2])
    assert result3 == True, f"테스트 3 실패: {result3}"
    print("테스트 3 통과!")
    
    print("\n모든 테스트 통과!")

test_contains_duplicate()
````

### 풀이 1: Set 활용 (가장 간단)
```python
def contains_duplicate(nums):
    return len(nums) != len(set(nums))
```
- `set(nums)`: 중복 제거
- 길이가 다르면 중복이 있었다는 뜻!

### 풀이 2: Set으로 하나씩 확인
```python
def contains_duplicate(nums):
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False
```
- 이미 본 숫자면 바로 True 반환
- 중복 발견 즉시 종료 (더 효율적일 수 있음)

### 핵심 포인트
- `set`은 중복을 허용하지 않음
- 원본 길이 vs set 길이 비교!
