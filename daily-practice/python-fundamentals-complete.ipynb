def my_sum(nums):

def filter_even(nums):

def find_indices(nums, target):

    for i, n in enumerate(nums):

def digit_sum(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

def first_last(lst):

# 5-2
def rotate_right(lst, k):
    return lst[-k:] + lst[:-k]

def count_chars(s):
    count = {}
    for c in s:
        count[c] = count.get(c, 0) + 1
    return count

# 6-2
def most_frequent(s):
    count = {}
    for c in s:
        count[c] = count.get(c, 0) + 1
    max_char = s[0]
    for c, cnt in count.items():
        if cnt > count[max_char]:
            max_char = c
    return max_char

# 6-3
def two_sum(nums, target):
    seen = {}
    for i, n in enumerate(nums):
        need = target - n
        if need in seen:
            return [seen[need], i]
        seen[n] = i
    return []

# 7-1
def has_duplicate(nums):
    return len(nums) != len(set(nums))

# 7-2
def intersection(nums1, nums2):
    return list(set(nums1) & set(nums2))

# 8-1
def is_palindrome(s):
    filtered = ''
    for c in s:
        if c.isalnum():
            filtered += c.lower()
    return filtered == filtered[::-1]

# 8-2
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
    count = {}
    for c in s1:
        count[c] = count.get(c, 0) + 1
    for c in s2:
        count[c] = count.get(c, 0) - 1
    return all(v == 0 for v in count.values())

# 9-1
def is_valid_parentheses(s):
    count = 0
    for c in s:
        if c == '(':
            count += 1
        elif c == ')':
            count -= 1
        if count < 0:
            return False
    return count == 0
