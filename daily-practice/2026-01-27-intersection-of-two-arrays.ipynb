## 문제: 두 배열의 교집합 (Intersection of Two Arrays)

### 난이도: Easy

### 왜 이 문제를 풀까?
- Set 연산 (교집합) 활용
- 두 배열 비교하는 패턴
- Python set 메서드 익히기

---

### 문제 설명

두 정수 배열 `nums1`과 `nums2`가 주어집니다.  
두 배열에 **공통으로 존재하는 숫자들**을 반환하세요.  
결과에서 각 숫자는 **한 번만** 나와야 합니다.

---

### 예제

**예제 1:**
```
입력: nums1 = [1, 2, 2, 1], nums2 = [2, 2]
출력: [2]
```

**예제 2:**
```
입력: nums1 = [4, 9, 5], nums2 = [9, 4, 9, 8, 4]
출력: [4, 9] 또는 [9, 4]
설명: 순서는 상관없음
```

---

### 제약 조건

- `1 <= nums1.length, nums2.length <= 1000`
- `0 <= nums1[i], nums2[i] <= 1000`

---

### 힌트

1. 두 배열을 set으로 바꾸면?
2. Set의 교집합 연산자: `&` 또는 `.intersection()`
3. 결과를 다시 list로 바꾸기: `list()`

````python
# def solution(nums1: list[int], nums2: list[int]) -> list[int]:
#     """
#     두 배열의 교집합을 반환합니다.
    
#     Args:
#         nums1: 첫 번째 정수 배열
#         nums2: 두 번째 정수 배열
    
#     Returns:
#         두 배열에 공통으로 있는 숫자들 (중복 제거)
#     """
#     # 여기에 코드를 작성하세요
#     pass

def solution(nums1: list[int], nums2: list[int]) -> list[int]:
    return list(set(nums1) & set(nums2))
````

---

## 테스트 케이스

````python
# 테스트 실행
def test_solution():
    # 테스트 케이스 1
    result1 = solution([1, 2, 2, 1], [2, 2])
    assert sorted(result1) == [2], f"테스트 1 실패: {result1}"
    print("테스트 1 통과!")
    
    # 테스트 케이스 2
    result2 = solution([4, 9, 5], [9, 4, 9, 8, 4])
    assert sorted(result2) == [4, 9], f"테스트 2 실패: {result2}"
    print("테스트 2 통과!")
    
    # 테스트 케이스 3: 교집합 없음
    result3 = solution([1, 2, 3], [4, 5, 6])
    assert result3 == [], f"테스트 3 실패: {result3}"
    print("테스트 3 통과!")
    
    print("\n모든 테스트 통과!")

test_solution()
````

### 풀이 1: Set 교집합 연산자 (가장 간단!)
```python
def solution(nums1, nums2):
    return list(set(nums1) & set(nums2))
```

### 풀이 2: intersection() 메서드
```python
def solution(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    return list(set1.intersection(set2))
```

### 풀이 3: for문으로 직접
```python
def solution(nums1, nums2):
    set2 = set(nums2)
    result = set()
    for num in nums1:
        if num in set2:
            result.add(num)
    return list(result)
```

### 핵심 포인트
- `set(nums)`: 중복 제거
- `&` 또는 `.intersection()`: 교집합
- `list()`: set을 list로 변환

### Set 연산 정리
| 연산 | 연산자 | 메서드 |
|------|--------|--------|
| 교집합 | `&` | `.intersection()` |
| 합집합 | `\|` | `.union()` |
| 차집합 | `-` | `.difference()` |
