## Part 1: 리스트 기본 연산

### 문제 1-1: 최댓값과 최솟값의 차
리스트에서 최댓값과 최솟값의 차이를 반환

````python
def max_min_diff(nums: list) -> int:
    return max(nums) - min(nums)

# 테스트
print(max_min_diff([1, 5, 10, 3, 8]))  # 9 (10-1)
print(max_min_diff([7, 7, 7]))         # 0
````

### 문제 1-2: 평균보다 큰 수의 개수
리스트의 평균값보다 큰 원소의 개수 반환

````python
def count_above_average(nums: list) -> int:
    # 힌트: 먼저 평균을 구하고, 평균보다 큰 개수 세기
    result = 0
    avg = sum(nums) / len(nums)
    for num in nums:
        if num > avg:
            result += 1
            
    return result

# 테스트
print(count_above_average([1, 2, 3, 4, 5]))  # 2 (4, 5가 평균 3.0보다 큼)
print(count_above_average([10, 10, 10]))     # 0
````

### 문제 1-3: 연속된 같은 숫자 제거
리스트에서 연속으로 나오는 중복 숫자를 하나만 남기기

````python
def remove_consecutive_duplicates(nums: list) -> list:
    # 힌트: 이전 숫자와 지금 숫자가 다를 때만 추가
    result = []
    for num in nums:
        if not result or num != result[-1]:
            result.append(num)
    return result 
# 테스트
print(remove_consecutive_duplicates([1, 1, 2, 2, 2, 3, 3, 1]))  # [1, 2, 3, 1]
print(remove_consecutive_duplicates([4, 4, 5, 5, 6]))           # [4, 5, 6]
````

## Part 2: 딕셔너리 심화

### 문제 2-1: 학생 성적 평균
딕셔너리로 받은 학생들의 성적에서 평균 구하기

````python
def average_score(scores: dict) -> float:
    # scores = {'Alice': 90, 'Bob': 85, 'Charlie': 95}
    return sum(scores.values()) / len(scores)

# 테스트
print(average_score({'Alice': 90, 'Bob': 85, 'Charlie': 95}))  # 90.0
print(average_score({'A': 100, 'B': 80}))                      # 90.0
````

### 문제 2-2: 키와 값 바꾸기
딕셔너리의 키와 값을 서로 바꾼 새 딕셔너리 반환

````python
def swap_dict(d: dict) -> dict:
    new_dict = {}
    for key, value in d.items():
        new_dict[value] = key

    return new_dict
# 와웅 맞췄다 개좋아 

# 테스트
print(swap_dict({'a': 1, 'b': 2, 'c': 3}))  # {1: 'a', 2: 'b', 3: 'c'}
print(swap_dict({'name': 'Kim', 'age': '25'}))  # {'Kim': 'name', '25': 'age'}
````

### 문제 2-3: 가장 많이 나온 단어
문장에서 가장 많이 등장한 단어 찾기

````python
def most_frequent_word(sentence: str) -> str:
    # 힌트: split()으로 단어 분리 후 딕셔너리로 개수 세기
    words = sentence.split()
    count = {}
    for word in words:
        if word in count:
            count[word] += 1
        else:
            count[word] = 1
    # for key, value in count.items():
    #     if value > count[-1]:

    max_word = words[0] # 초기값 지정 첫단어를 일단 최대라 지정
    for word, cnt in count.items():
        if cnt > count[max_word]:
            max_word = word

    return max_word
# 테스트
print(most_frequent_word("apple banana apple cherry apple"))  # "apple"
print(most_frequent_word("hello world hello"))                # "hello"
````

## Part 3: 문자열 조작

### 문제 3-1: 대소문자 반전
대문자는 소문자로, 소문자는 대문자로 변환

````python
def swap_case(s: str) -> str:
    # 힌트: char.isupper(), char.islower(), char.upper(), char.lower()
    result = ""
    for char in s:
        if char.isupper():
            result += char.lower()
        elif char.islower():
            result += char.upper()
        else:
            result += char
    return result 

# 테스트
print(swap_case("Hello World"))  # "hELLO wORLD"
print(swap_case("PyThOn"))       # "pYtHoN"
````

### 문제 3-2: 회문(팰린드롬) 만들기
문자열이 회문인지 확인 (앞으로 읽으나 뒤로 읽으나 같은지)

````python
def is_palindrome(s: str) -> bool:
    # 힌트: 뒤집은 문자열과 비교 ([::-1] 사용 가능)
    if s[::] == s[::-1]:
        return True
    else:
        return False


# 테스트
print(is_palindrome("level"))   # True
print(is_palindrome("hello"))   # False
print(is_palindrome("noon"))    # True
````

### 문제 3-3: 첫 글자만 대문자
각 단어의 첫 글자만 대문자로 만들기

````python
def capitalize_words(sentence: str) -> str:
    # 힌트: split()으로 단어 분리, 각 단어의 [0]만 대문자
    result = []
    div_word = sentence.split()
    for char in div_word:
        result.append(char[0].upper() + char[1:])

    return " ".join(result)
# 테스트
print(capitalize_words("hello world python"))  # "Hello World Python"
print(capitalize_words("i love coding"))       # "I Love Coding"
````

## Part 4: enumerate & 인덱스 활용

### 문제 4-1: 홀수 인덱스 합
인덱스가 홀수(1, 3, 5...)인 원소들의 합

````python
def sum_odd_indices(nums: list) -> int:
    result  = 0
    for index, value in enumerate(nums):
        if index % 2 == 1:
            result += value
    return result 
# 테스트
print(sum_odd_indices([10, 20, 30, 40, 50]))  # 60 (20+40)
print(sum_odd_indices([1, 2, 3, 4]))          # 6 (2+4)
````

### 문제 4-2: 값과 인덱스가 같은 원소
리스트에서 "값 == 인덱스"인 원소들만 반환

````python
def match_index_value(nums: list) -> list:
    result = []
    for index, value in enumerate(nums):
        if index == value:
            result.append(value)
    return result

# 테스트
print(match_index_value([0, 1, 5, 3, 4]))  # [0, 1, 3, 4]
print(match_index_value([5, 6, 7]))        # []
````

## Part 5: while문 심화

### 문제 5-1: 숫자 뒤집기
정수를 뒤집은 결과 반환 (예: 1234 → 4321)

````python
def reverse_number(n: int) -> int:
    # 힌트: n % 10으로 마지막 자리 추출, result = result * 10 + 마지막자리
    result = 0 
    # last_num = n % 10

    while n > 0:
        last_dight = n % 10
        result = result * 10 + last_dight
        n = n //10
        
    return result 

# 테스트
print(reverse_number(1234))  # 4321
print(reverse_number(567))   # 765
````

### 문제 5-2: 제곱수 찾기
n 이하의 완전제곱수 개수 (1, 4, 9, 16, ...)

````python
def count_squares(n: int) -> int:
    # 힌트: i*i가 n 이하인 동안 카운트
    count = 0
    i = 1

    while 1 * i <= n:
        count  += 1
        i += 1
    return count 
        

# 테스트
print(count_squares(10))  # 3 (1, 4, 9)
print(count_squares(25))  # 5 (1, 4, 9, 16, 25)
````

## Part 6: 종합 응용 문제

### 문제 6-1: 두 리스트 교집합
두 리스트의 공통 원소만 반환 (중복 제거)

````python
def intersection(list1: list, list2: list) -> list:
    # 힌트: set 활용, & 연산자
    return list(set(list1) & set(list2))

# 테스트
print(sorted(intersection([1, 2, 3], [2, 3, 4])))  # [2, 3]
print(sorted(intersection([1, 1, 2], [2, 2, 3])))  # [2]
````

### 문제 6-2: 문자열 압축
연속된 문자를 "문자+개수" 형태로 압축

````python
def compress_string(s: str) -> str:
    # 힌트: 이전 문자와 비교하며 카운트
    if not s:
        return ""
    
    result = ""
    current_char = s[0]
    count = 1

    for i in range(1, len(s)):
        if s[i] == current_char:
            count += 1
        else:
            result += current_char + str(count)
            current_char = s[i]
            count = 1
    result += current_char + str(count)

    return result 


# 테스트
print(compress_string("aaabbbbcc"))  # "a3b4c2"
print(compress_string("abc"))        # "a1b1c1"
````

### 문제 6-3: 두 번째로 큰 수
리스트에서 두 번째로 큰 수 찾기

````python
def second_largest(nums: list) -> int:
    # 힌트: 중복 제거 후 정렬, 또는 최댓값 찾고 제거한 뒤 다시 최댓값
    result = []
    for num in nums:
        if num not in result:
            result.append(num)
    
    result.sort()
    return result[-2]

# 테스트
print(second_largest([1, 5, 3, 9, 2]))  # 5
print(second_largest([10, 10, 8, 7]))   # 8
````
