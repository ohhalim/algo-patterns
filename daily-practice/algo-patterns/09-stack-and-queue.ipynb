# 09. Stack and Queue

괄호 매칭, 최근 요청 유지, 단조 스택 문제를 위한 기본 패턴.


## 패턴 신호
- "최근 것부터 처리" -> Stack
- "먼저 들어온 순서대로 처리" -> Queue
- "다음 더 큰/작은 원소" -> Monotonic Stack

## 브루트포스 -> 최적화
1. 각 원소마다 오른쪽을 끝까지 탐색하면 O(n^2)
2. 스택에 후보를 저장해서 한 번 순회 O(n)


```python
from collections import deque

# 기본 사용 예시
stack = []
stack.append(10)
stack.append(20)
print(stack.pop())  # 20

queue = deque()
queue.append(10)
queue.append(20)
print(queue.popleft())  # 10
```
