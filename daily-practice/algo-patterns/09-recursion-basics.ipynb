## 핵심 개념

### 재귀의 3요소
1. **Base Case** (종료 조건) - 재귀를 멈추는 조건. 이게 없으면 무한 루프!
2. **Recursive Case** (재귀 호출) - 문제를 더 작은 문제로 쪼개기
3. **Return** - 결과를 합치기

### 기본 템플릿

## 예제: 팩토리얼

````python
# 예제 - 이해하고 넘어가세요
def factorial_example(n):
    if n <= 1:          # Base case
        return 1
    return n * factorial_example(n - 1)  # Recursive case

print(factorial_example(5))  # 120 (5*4*3*2*1)
````
