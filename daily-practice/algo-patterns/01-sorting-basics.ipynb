# 01. 정렬 (Sorting) 기초

정렬은 모든 알고리즘의 기본. 직접 구현해보면 반복문/조건문/swap이 완전히 손에 익는다.

---

## 핵심 개념

| 정렬 | 시간복잡도 | 특징 |
|------|-----------|------|
| Bubble Sort | O(n^2) | 인접한 두 원소 비교/교환 |
| Selection Sort | O(n^2) | 최솟값을 찾아서 앞으로 |
| Insertion Sort | O(n^2) | 카드 정렬처럼 삽입 |
| Python sorted() | O(n log n) | Timsort, 실전에서 이걸 씀 |

## swap 패턴 (파이썬)
```python
a, b = b, a  # 파이썬에서는 이렇게 간단!
```

## 예제: Bubble Sort 구현

인접한 두 원소를 비교해서 큰 값을 뒤로 보내는 방식

````python
# 버블 정렬(Bubble Sort)
# 핵심 아이디어:
# "인접한 두 값을 비교해서, 큰 값을 오른쪽으로 계속 밀어낸다."
# 바깥 반복 1번이 끝날 때마다 '가장 큰 값 1개'가 맨 뒤에 고정된다.

def bubble_sort_example(nums):
    # nums는 숫자들이 들어 있는 리스트(숫자 줄)야.
    # 예: [64, 34, 25, 12, 22, 11, 90]

    n = len(nums)  # 숫자가 총 몇 개인지 센다.

    # i는 "정리 라운드 번호"야.
    # 한 라운드가 끝날 때마다 가장 큰 숫자 1개가 맨 뒤에 자리 잡는다.
    for i in range(n):

        # 맨 뒤 i개는 이미 정리 완료라서 다시 비교하지 않는다.
        # 그래서 앞쪽 구간만 비교한다.
        for j in range(0, n - i - 1):

            # j번째 숫자와 j+1번째 숫자를 비교한다.
            # 왼쪽 숫자가 더 크면 순서가 뒤집힌 상태!
            if nums[j] > nums[j + 1]:
                # 두 숫자의 자리를 바꿔서 작은 숫자를 왼쪽으로 보낸다.
                nums[j], nums[j + 1] = nums[j + 1], nums[j]

    # 모든 라운드가 끝나면 작은 수 -> 큰 수 순서로 정렬된다.
    return nums

print(bubble_sort_example([64, 34, 25, 12, 22, 11, 90]))
# [11, 12, 22, 25, 34, 64, 90]


def bubble_sort_example(nums):
    # nums는 숫자들이 들어있는 리스트(숫자 줄) 야.

    n = len(nums) # 숫자가 총 몇개인지 센다

    # i는 정리 라운드 번호야
    # 한라운드가 끝날때마다 가장큰숫자 1개가 맨뒤에 자리잡는다
    for i in range(n): 
        # 맨뒤 i개는 이미 정리완료라서 다시 비교하지 않는다 
        # 그레서 앞쪽 구간만 비교한다
        for j in range(0, n -i -1):

            #  j번째 숫자와 j+1번째 숫자를 비교한다
            # 왼쪽숫자가 더 크면 순서가 뒤집흰 상태다
            if num[j] > num[j + 1]:
                # 두 숫자의 자리를 바꿔서 작은 숫자를 왼쪽으로 보낸다
                nums[j], nums[j+1] = nums[j +1], nums[j]

        return nums 
print(bubble_sort_example([64, 34, 25, 12, 22, 11, 90]))

def bubble_sort(nums):
    # 숫자가 몇 개인지 세는거야
    n = len(nums)

    # 같은 작업을 여러번 반복해
    # 왜 반복하냐면 , 한번에 다 정리되지 않기 때문이야 
    for i in range(n):
        # 앞에서 부터 옆 숫자끼리 비요해
        # 뒤쪽은 이미 정리된 자리라서 점점 비교를 안해도 돼
        for j in range(0, n -i -1):
            # 왼쪽 숫자가 오른쪽 숫자보다 크면 순서가 틀린거야 
            if nums[j] > nums[j+1]:
                #둘의 자리를 바꿔 
                # 작은 수자는 왼쪽으로 큰숫자는 오른쪽으로
                nums[j], nums[j+1] = nums[j+1], nums[j]
    return nums
    # 이걸 한바퀴 돌면
    # 가장큰 숫자 하나가 맨오른쪽에 확정돼
    # 다시 돌면?
    # 가장큰숫자 하나가 맨오른쪽에 확정돼
    # 계속 반복하면 결국 작은숫자부터 큰숫자까지 다 정렬돼 
    # 비눗방울이 물위로 올라오듯, 큰숫자가 오른쪽끝으로 떠오르는 정렬이라서 버블정렬이야

def bubble_sort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(0, n-i -1):
            if  nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j +1], nums[j]
    return nums
    
````

## 문제 1: Selection Sort 구현
배열에서 최솟값을 찾아 맨 앞과 교환하는 정렬

**힌트**: 바깥 루프에서 i번째 위치에 올 최솟값을 찾는다

````python
def selection_sort(nums):
    # 여기에 코드를 작성하세요
    n = len(nums)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if nums[j] < nums[min_idx]:
                min_idx =j
        nums[i], nums[min_idx] = nums[min_idx], nums[i]

    return nums
# 테스트
print(selection_sort([64, 25, 12, 22, 11]))  # [11, 12, 22, 25, 64]
print(selection_sort([5, 1, 4, 2, 8]))       # [1, 2, 4, 5, 8]
print(selection_sort([1]))                    # [1]
print(selection_sort([3, 3, 1, 1, 2]))       # [1, 1, 2, 3, 3]
````
