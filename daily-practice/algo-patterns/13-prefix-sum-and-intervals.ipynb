# 13. Prefix Sum and Intervals

누적합으로 구간 합을 빠르게 처리하고, 구간 병합 패턴을 익힙니다.


## 패턴 신호
- "연속 구간 합"
- "합이 k인 부분 배열 개수"
- "겹치는 구간 병합"

## 브루트포스 -> 최적화
1. 모든 구간을 직접 더하면 O(n^2)
2. prefix sum + hashmap으로 O(n)
3. interval은 시작점 정렬 후 한 번 순회 O(n log n)


```python
# 누적합 기본 예시
nums = [3, 1, 4, 1, 5]
prefix = [0]
for x in nums:
    prefix.append(prefix[-1] + x)

# nums[l:r+1] 합 = prefix[r+1] - prefix[l]
print(prefix[4] - prefix[1])  # 1+4+1 = 6
```
