# 04. 빈도수 카운팅 (Frequency Counting)

딕셔너리(HashMap)로 개수를 세는 패턴. 면접에서 가장 많이 나오는 패턴 중 하나!

---

## 핵심 패턴

### 기본 카운팅
```python
count = {}
for item in lst:
    count[item] = count.get(item, 0) + 1
```

### collections.Counter 활용
```python
from collections import Counter
count = Counter(lst)           # 바로 딕셔너리 생성
count.most_common(3)           # 가장 많은 3개
```

### defaultdict 활용
```python
from collections import defaultdict
d = defaultdict(int)          # 기본값 0
d = defaultdict(list)         # 기본값 []
```

## 예제: 문자열에서 가장 많이 나온 문자

````python
# 예제 - 이해하고 넘어가세요
def most_common_char_example(s):
    count = {}
    for char in s:
        count[char] = count.get(char, 0) + 1
    
    max_char = ''
    max_count = 0
    for char, cnt in count.items():
        if cnt > max_count:
            max_count = cnt
            max_char = char
    return max_char

print(most_common_char_example("aabbbcc"))  # 'b'
````
