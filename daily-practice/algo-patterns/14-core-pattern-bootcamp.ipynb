## 1) Hash
**신호**: 중복/빈도/쌍 찾기, O(n^2) 줄이고 싶을 때

**암기 카드**
- 한 번 순회하면서 필요한 정보를 dict/set에 저장
- `찾아야 할 값`을 바로 계산해서 조회


```python
# 문제: Two Sum
# nums에서 합이 target이 되는 두 인덱스를 반환
def two_sum(nums: list[int], target: int):
    # TODO: Hash 패턴으로 O(n)에 구현
    seen = {} # 초기값 생성
    for i, num in enumerate(nums): # nums 인덱스 i 와 값 num으로 하나씩 순회
        need = target - num # 현재숫자 num과 더해서 target이 되려면 필요한값 need 계산
        if need in seen: # 필요한 값 need를 이전에 본적이 있다면
            return [seen[need], i] # need 이전 인덱스와 현재 인덱스 i를 정답으로 즉시 반환
        seen[num] = i # 아직 정답을 못찾았다면 현재 숫자 num의 인덱스 i 을 딕셔너리에 기록
    return [] # 문제 조건상 보통 실행되지 않지만 정답이 없을 때를 대비해 빈리스트 반환 

def two_sum(nums:list[int], target: int):
    seen = {}
    for i, num in enumerate(nums):
        need = target - num
        if need in seen:
            return [seen[need], i]
        seen[num] = i
    return []


check('two_sum #1', two_sum([2, 7, 11, 15], 9), [0, 1])
check('two_sum #2', two_sum([3, 2, 4], 6), [1, 2])
check('two_sum #3', two_sum([3, 3], 6), [0, 1])
```

## 2) Two Pointers
**신호**: 정렬 배열, 양끝 비교, 부분 조건 만족

**암기 카드**
- left/right 포인터로 범위를 좁힌다
- 포인터 이동 규칙을 먼저 정한다


```python
# 문제: Valid Palindrome
# 영숫자만 남기고 대소문자 무시했을 때 팰린드롬인지 판단
def is_palindrome(s):
    # TODO: Two pointers로 구현
    left, right = 0, len(s) - 1
    while left < right: 
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    return True

        


check('is_palindrome #1', is_palindrome('A man, a plan, a canal: Panama'), True)
check('is_palindrome #2', is_palindrome('race a car'), False)
check('is_palindrome #3', is_palindrome(' '), True)
```
