# 02. 탐색 (Searching) 패턴

데이터에서 원하는 값을 찾는 가장 기본적인 알고리즘

---

## 핵심 개념

| 탐색 | 시간복잡도 | 조건 |
|------|-----------|------|
| Linear Search | O(n) | 제한 없음 |
| Binary Search | O(log n) | **정렬된 배열** 필수 |

## Binary Search 핵심 패턴
```python
left, right = 0, len(nums) - 1
while left <= right:
    mid = (left + right) // 2
    if nums[mid] == target:
        return mid
    elif nums[mid] < target:
        left = mid + 1
    else:
        right = mid - 1
return -1  # 못 찾음
```

---
## 예제: Linear Search

````python
# 예제 - 이해하고 넘어가세요
def linear_search_example(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1

print(linear_search_example([4, 2, 7, 1, 9], 7))  # 2
print(linear_search_example([4, 2, 7, 1, 9], 5))  # -1
````
