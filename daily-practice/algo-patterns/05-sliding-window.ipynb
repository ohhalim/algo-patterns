# 05. 슬라이딩 윈도우 (Sliding Window)

연속 구간 문제를 O(n)으로 해결하는 패턴. 부분 배열/부분 문자열 문제에서 필수!

---

## 핵심 패턴 2가지

### 1. 고정 크기 윈도우
```python
# 크기 k인 윈도우를 슬라이딩
window_sum = sum(nums[:k])  # 첫 윈도우
max_sum = window_sum
for i in range(k, len(nums)):
    window_sum += nums[i] - nums[i - k]  # 새 원소 추가, 오래된 원소 제거
    max_sum = max(max_sum, window_sum)
```

### 2. 가변 크기 윈도우
```python
left = 0
for right in range(len(s)):
    # 윈도우에 s[right] 추가
    while 윈도우가_조건을_위반:
        # 윈도우에서 s[left] 제거
        left += 1
    # 결과 갱신
```

## 예제: 크기 K 부분 배열의 최대 합

````python
# 예제 - 이해하고 넘어가세요
def max_sum_subarray_example(nums, k):
    window_sum = sum(nums[:k])
    max_sum = window_sum
    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]
        max_sum = max(max_sum, window_sum)
    return max_sum

print(max_sum_subarray_example([2, 1, 5, 1, 3, 2], 3))  # 9 (5+1+3)
````
