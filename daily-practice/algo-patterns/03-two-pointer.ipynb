# 03. 투 포인터 (Two Pointer)

O(n^2)를 O(n)으로 줄이는 마법. 배열 문제의 핵심 패턴!

---

## 핵심 패턴 2가지

### 1. 양 끝에서 좁혀오기
```python
left, right = 0, len(nums) - 1
while left < right:
    # 조건에 따라 left++ 또는 right--
```

### 2. 같은 방향 (Slow/Fast)
```python
slow = 0
for fast in range(len(nums)):
    if 조건:
        nums[slow] = nums[fast]
        slow += 1
```

---
## 예제: 정렬된 배열의 두 수 합

정렬된 배열에서 합이 target인 두 수의 인덱스 찾기

````python
# 예제 - 이해하고 넘어가세요
def two_sum_sorted_example(nums, target):
    left, right = 0, len(nums) - 1
    while left < right:
        total = nums[left] + nums[right]
        if total == target:
            return [left, right]
        elif total < target:
            left += 1
        else:
            right -= 1
    return [-1, -1]

print(two_sum_sorted_example([2, 7, 11, 15], 9))  # [0, 1]
````
