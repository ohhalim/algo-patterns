## 예제: 주식 매매 최적 시점

```python
# 예제 - 이해하고 넘어가세요
def max_profit_example(prices):
    min_price = float('inf')
    max_profit = 0
    for price in prices:
        min_price = min(min_price, price)       # 지금까지 최저가
        max_profit = max(max_profit, price - min_price)  # 지금 팔면 이익
    return max_profit

print(max_profit_example([7, 1, 5, 3, 6, 4]))  # 5 (1에 사서 6에 팔기)
```
