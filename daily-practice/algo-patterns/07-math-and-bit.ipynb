## 핵심 개념

### 자주 쓰는 수학 연산
```python
n % 10        # 마지막 자릿수
n // 10       # 마지막 자릿수 제거
n % 2 == 0    # 짝수 판별
abs(n)        # 절대값
divmod(n, d)  # (몫, 나머지) 동시에
```

### 비트 연산
```python
a & b    # AND: 둘 다 1이면 1
a | b    # OR: 하나라도 1이면 1  
a ^ b    # XOR: 다르면 1 (같은 수 XOR하면 0!)
~a       # NOT: 비트 반전
a << n   # 왼쪽 시프트 (x2^n)
a >> n   # 오른쪽 시프트 (/2^n)
```

### XOR의 마법
- `x ^ 0 = x`
- `x ^ x = 0`
- XOR은 교환법칙, 결합법칙 성립
