## 자주 쓰는 문자열 메서드

```python
s.lower()           # 소문자 변환
s.upper()           # 대문자 변환
s.strip()           # 양쪽 공백 제거
s.split()           # 공백 기준 분리 -> 리스트
' '.join(lst)       # 리스트 -> 문자열

s.isalpha()         # 알파벳만?
s.isdigit()         # 숫자만?
s.isalnum()         # 영숫자?

ord('a')            # 아스키 코드 (97)
chr(97)             # 문자 ('a')
```
