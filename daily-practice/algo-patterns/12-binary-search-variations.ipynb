이분 탐색의 핵심은 `정답 구간`을 절반씩 줄이는 것입니다.


## 패턴 신호
- 정렬된 배열
- "최소/최대 인덱스", "삽입 위치"
- O(log n) 요구

## 브루트포스 -> 최적화
1. 선형 탐색은 O(n)
2. 중간값 비교로 탐색 구간 절반 제거 O(log n)


```python
# 기본 템플릿
def binary_search_template(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        if nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```
