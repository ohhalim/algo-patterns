## 문제: 유효한 애너그램 (Valid Anagram)

### 난이도: Easy

---

### 문제 설명

두 문자열 `s`와 `t`가 주어집니다.  
`t`가 `s`의 **애너그램**이면 `True`, 아니면 `False`를 반환하세요.

**애너그램**: 문자를 재배열해서 다른 단어를 만들 수 있는 것
- 예: "anagram" -> "nagaram" (같은 문자들로 구성)

---

### 예제

**예제 1:**
```
입력: s = "anagram", t = "nagaram"
출력: True
```

**예제 2:**
```
입력: s = "rat", t = "car"
출력: False
```

**예제 3:**
```
입력: s = "a", t = "a"
출력: True
```

---

### 제약 조건

- `1 <= s.length, t.length <= 5 * 10^4`
- `s`와 `t`는 소문자 알파벳으로만 구성됨

---

### 힌트

1. 두 문자열의 길이가 다르면? -> 바로 False
2. 각 문자의 개수를 세서 비교하면?
3. Python `sorted()` 함수도 활용 가능!
4. `collections.Counter` 도 유용해요

````python
def is_anagram(s: str, t: str) -> bool:
    """
    t가 s의 애너그램인지 확인합니다.
    
    Args:
        s: 첫 번째 문자열
        t: 두 번째 문자열
    
    Returns:
        애너그램이면 True, 아니면 False
    """
    # 여기에 코드를 작성하세요
    if sorted(s) == sorted(t):
        return True
    else:
        return False
````

---

## 테스트 케이스

````python
# 테스트 실행
def test_is_anagram():
    # 테스트 케이스 1
    result1 = is_anagram("anagram", "nagaram")
    assert result1 == True, f"테스트 1 실패: {result1}"
    print("테스트 1 통과!")
    
    # 테스트 케이스 2
    result2 = is_anagram("rat", "car")
    assert result2 == False, f"테스트 2 실패: {result2}"
    print("테스트 2 통과!")
    
    # 테스트 케이스 3
    result3 = is_anagram("a", "a")
    assert result3 == True, f"테스트 3 실패: {result3}"
    print("테스트 3 통과!")
    
    # 테스트 케이스 4: 길이가 다른 경우
    result4 = is_anagram("ab", "a")
    assert result4 == False, f"테스트 4 실패: {result4}"
    print("테스트 4 통과!")
    
    print("\n모든 테스트 통과!")

test_is_anagram()
````

### 풀이 1: 정렬 비교 (간단)
```python
def is_anagram(s, t):

```
- 시간복잡도: O(n log n)

### 풀이 2: Counter 사용
```python
from collections import Counter

def is_anagram(s, t):
    return Counter(s) == Counter(t)
```
- 시간복잡도: O(n)
- 공간복잡도: O(n)

### 풀이 3: HashMap 직접 구현
```python
def is_anagram(s, t):
    if len(s) != len(t):
        return False
    
    count = {}
    for c in s:
        count[c] = count.get(c, 0) + 1
    
    for c in t:
        if c not in count or count[c] == 0:
            return False
        count[c] -= 1
    
    return True
```

- 애너그램 = 같은 문자 구성, 순서만 다름
- 정렬하면 같아져야 함
- 또는 각 문자 개수가 같아야 함
