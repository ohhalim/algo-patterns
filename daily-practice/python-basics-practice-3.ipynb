## Part 1: 리스트 기초

### 문제 1-1: 리스트 첫 번째와 마지막
리스트의 첫 번째와 마지막 값을 리스트로 반환

````python
def first_and_last(lst: list) -> list:
    # result = []
    # first_and_last = lst[0] + lst[-1]
    # result.append.first_and_last
    # return result

    return [lst[0], lst[-1]]
# 테스트
print(first_and_last([1, 2, 3, 4, 5]))  # [1, 5]
print(first_and_last(['a', 'b', 'c']))  # ['a', 'c']
````

### 문제 1-2: 리스트 평균
숫자 리스트의 평균 반환

````python
def average(nums: list) -> float:
    sum_list = 0
    for num in nums:
        sum_list += num
    result = sum_list / len(nums)
    return result 
# 테스트
print(average([1, 2, 3, 4, 5]))  # 3.0
print(average([10, 20]))         # 15.0
````

### 문제 1-3: 양수만 필터
양수만 담은 리스트 반환

````python
def filter_positive(nums: list) -> list:
    int_list = [] 
    for num in nums:
        if num > 0:
            int_list.append(num)
    return int_list

# 와 개젛아 기분

# 테스트
print(filter_positive([1, -2, 3, -4, 5]))  # [1, 3, 5]
print(filter_positive([-1, -2, -3]))       # []
````

## Part 2: 문자열 기초

### 문제 2-1: 공백 제거
문자열에서 공백 제거

````python
def remove_spaces(s: str) -> str:
    # return s.join( )
    a = s.split()
    result = str()
    for i in a:
        result += i
    return result
# 테스트
print(remove_spaces('hello world'))   # 'helloworld'
print(remove_spaces('a b c d'))       # 'abcd'
````

### 문제 2-2: 대문자 개수
문자열에서 대문자 개수 반환

````python
def count_uppercase(s: str) -> int:
    # 힌트: char.isupper() 사용
    result = 0
    for char in s:
        if char.isupper():
            result += 1
    return result 

# 테스트
print(count_uppercase('Hello World'))  # 2
print(count_uppercase('ABC'))          # 3
print(count_uppercase('abc'))          # 0
````

### 문제 2-3: 첫 글자 대문자로
각 단어의 첫 글자를 대문자로

````python
def capitalize_words(s: str) -> str:
    # 힌트: split, upper(), join
    words = s.split()
    # upper_word = ""
    # for char in words:
    #    upper_word += char[0].upper()
    # return upper_word.join()
    capitalized = []
    for word in words:
        capitalized.append(word[0].upper()+ word[1:])
    return " ".join(capitalized)
# 테스트
print(capitalize_words('hello world'))  # 'Hello World'
print(capitalize_words('python is fun'))  # 'Python Is Fun'
````

## Part 3: 딕셔너리 기초

### 문제 3-1: 값 두 배로
딕셔너리의 모든 값을 2배로

````python
def double_values(d: dict) -> dict:
    # enumerate은 인덱스 랑 값 순회 하는거고 #item 은 딕셔너리 키 값순회 하는거
    result = {}
    # for key, value in d.items[key, vaule]:
    #     result += key, value * 2
    # return result 
    for key, value in d.items():
        result[key] = value *2 # 아 그 키에 값을 추가하는 방식으로 하는구나 
    return result 
# 테스트
print(double_values({'a': 1, 'b': 2}))  # {'a': 2, 'b': 4}
print(double_values({'x': 5}))          # {'x': 10}
````

### 문제 3-2: 최대 값의 키
가장 큰 값을 가진 키 반환

````python
def key_of_max_value(d: dict) -> str:
    # result = {"", 0}
    # for key, value in d.items():
    #     if value > result[key]:
    #         result = {key, value}
    # return key 
    max_key = ""
    max_value = 0
    for key, value in d.items():
        if value > max_value:
            max_key = key
            max_value =value
    return max_key
# 테스트
print(key_of_max_value({'a': 1, 'b': 5, 'c': 3}))  # 'b'
print(key_of_max_value({'x': 10, 'y': 2}))         # 'x'
````

### 문제 3-3: 두 딕셔너리 합치기
두 딕셔너리를 하나로 합치기

````python
def merge_dicts(d1, d2):
    result = {}
    for key, value in d1.items():
        result[key] = value
    for key, value in d2.items():
        result[key] = value
    return result 

# 테스트
print(merge_dicts({'a': 1}, {'b': 2}))  # {'a': 1, 'b': 2}
print(merge_dicts({'x': 1, 'y': 2}, {'z': 3}))  # {'x': 1, 'y': 2, 'z': 3}
````

## Part 4: 조건문 심화

### 문제 4-1: 3의 배수 또는 5의 배수
3의 배수 또는 5의 배수인지 확인

````python
def is_multiple_of_3_or_5(n):
    return n % 3 == 0 or n % 5 == 0

# 테스트
print(is_multiple_of_3_or_5(9))   # True (3의 배수)
print(is_multiple_of_3_or_5(10))  # True (5의 배수)
print(is_multiple_of_3_or_5(7))   # False
````

### 문제 4-2: 윤년 판별
윤년인지 확인 (4의 배수, 단 100의 배수는 제외, 400의 배수는 포함)

````python
def is_leap_year(year):
    if year % 400 == 0:
        return True
    if year % 100 == 0:
        return False
    if year % 4 == 0:
        return True
    return False
    

# 테스트
print(is_leap_year(2020))  # True
print(is_leap_year(1900))  # False (100의 배수)
print(is_leap_year(2000))  # True (400의 배수)
````

## Part 5: 종합

### 문제 5-1: 두 번째로 큰 값
리스트에서 두 번째로 큰 값 반환

````python
def second_largest(nums):
    first = nums[0]
    second = None
    for n in nums:
        if n > first:
            second = first
            first = n
        elif second is None or n > second:
            if n != first:
                second = n
    return second
             

# 테스트
print(second_largest([1, 5, 2, 8, 3]))  # 5
print(second_largest([10, 20, 30]))     # 20
````

### 문제 5-2: 중복 체크
리스트에 중복이 있는지 확인

````python
def has_duplicates(lst):
    return len(lst) != len(set(lst))

# 테스트
print(has_duplicates([1, 2, 3, 2]))  # True
print(has_duplicates([1, 2, 3]))     # False
````
