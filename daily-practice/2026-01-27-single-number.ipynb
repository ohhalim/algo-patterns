## 문제: 싱글 넘버 (Single Number)

### 난이도: Easy

### 왜 이 문제를 풀까?
- Set 활용 복습
- XOR 비트 연산 기초 학습
- "중복 제거" 패턴의 다른 접근법

---

### 문제 설명

정수 배열 `nums`가 주어집니다.  
배열에서 **모든 숫자는 두 번씩 나오고, 딱 하나만 한 번 나옵니다.**  
그 **한 번만 나오는 숫자**를 찾아서 반환하세요.

---

### 예제

**예제 1:**
```
입력: nums = [2, 2, 1]
출력: 1
설명: 2는 두 번, 1은 한 번
```

**예제 2:**
```
입력: nums = [4, 1, 2, 1, 2]
출력: 4
설명: 1과 2는 두 번씩, 4는 한 번
```

**예제 3:**
```
입력: nums = [1]
출력: 1
```

---

### 제약 조건

- `1 <= nums.length <= 3 * 10^4`
- `-3 * 10^4 <= nums[i] <= 3 * 10^4`
- 각 요소는 두 번 나오거나, 딱 한 번만 나옴

---

### 힌트

1. Set으로 "처음 보면 추가, 또 보면 제거" 하면?
2. 또는: 각 숫자의 개수를 세서 1인 것 찾기
3. 고급: XOR 연산 (같은 숫자끼리 XOR하면 0이 됨!)

````python
# def single_number(nums: list[int]) -> int:
#     """
#     배열에서 한 번만 나오는 숫자를 찾습니다.
    
#     Args:
#         nums: 정수 배열 (한 숫자만 한 번, 나머지는 두 번씩)
    
#     Returns:
#         한 번만 나오는 숫자
#     """
#     # 여기에 코드를 작성하세요
    
def solution(nums: list[int]) -> int:
    seen = set()
    for num in nums:
        if num in seen:
            seen.remove(num)
        else:
            seen.add(num)
    return seen.pop() # 
````

---

## 테스트 케이스

````python
# 테스트 실행
def test_single_number():
    # 테스트 케이스 1
    result1 = solution([2, 2, 1])
    assert result1 == 1, f"테스트 1 실패: {result1}"
    print("테스트 1 통과!")
    
    # 테스트 케이스 2
    result2 = solution([4, 1, 2, 1, 2])
    assert result2 == 4, f"테스트 2 실패: {result2}"
    print("테스트 2 통과!")
    
    # 테스트 케이스 3
    result3 = solution([1])
    assert result3 == 1, f"테스트 3 실패: {result3}"
    print("테스트 3 통과!")
    
    print("\n모든 테스트 통과!")

test_single_number()
````

### 풀이 1: Set 활용
```python
def single_number(nums):
    seen = set()
    for num in nums:
        if num in seen:
            seen.remove(num)  # 두 번째 보면 제거
        else:
            seen.add(num)     # 처음 보면 추가
    return seen.pop()  # 남은 하나가 정답
```

### 풀이 2: 수학 공식
```python
def single_number(nums):
    return 2 * sum(set(nums)) - sum(nums)
```
- `set(nums)`의 합 × 2 = 모든 숫자가 두 번씩 나온다고 가정한 합
- 실제 합을 빼면 = 한 번만 나온 숫자

### 풀이 3: XOR 비트 연산 (가장 효율적!)
```python
def single_number(nums):
    result = 0
    for num in nums:
        result ^= num  # XOR 연산
    return result
```
- `a ^ a = 0` (같은 숫자끼리 XOR하면 0)
- `0 ^ a = a` (0과 XOR하면 자기 자신)
- 두 번 나오는 숫자들은 서로 상쇄되고, 한 번만 나오는 숫자만 남음!

### 핵심 포인트
- Set: 추가/제거로 중복 처리
- XOR: 같은 값은 상쇄됨
