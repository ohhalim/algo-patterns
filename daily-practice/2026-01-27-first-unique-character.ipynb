## 문제: 문자열의 첫 번째 고유 문자 (First Unique Character)

### 난이도: Easy

### 왜 이 문제를 풀까?
- HashMap으로 빈도수 세기
- 문자열 순회 패턴
- "처음 조건 만족하는 것 찾기" 패턴

---

### 문제 설명

문자열 `s`가 주어집니다.  
**처음으로 한 번만 나오는 문자의 인덱스**를 반환하세요.  
그런 문자가 없으면 `-1`을 반환하세요.

---

### 예제

**예제 1:**
```
입력: s = "leetcode"
출력: 0
설명: 'l'이 처음으로 한 번만 나오는 문자 (인덱스 0)
```

**예제 2:**
```
입력: s = "loveleetcode"
출력: 2
설명: 'v'가 처음으로 한 번만 나오는 문자 (인덱스 2)
```

**예제 3:**
```
입력: s = "aabb"
출력: -1
설명: 모든 문자가 두 번씩 나옴
```

---

### 제약 조건

- `1 <= s.length <= 10^5`
- `s`는 소문자 영어 알파벳으로만 구성

---

### 힌트

1. 먼저 각 문자가 몇 번 나오는지 세기 (HashMap)
2. 다시 문자열을 순회하면서 개수가 1인 첫 번째 문자 찾기
3. `collections.Counter` 또는 dict 사용

````python
# def solution(s: str) -> int:
#     """
#     처음으로 한 번만 나오는 문자의 인덱스를 반환합니다.
    
#     Args:
#         s: 입력 문자열
    
#     Returns:
#         첫 번째 고유 문자의 인덱스, 없으면 -1
#     """
#     # 여기에 코드를 작성하세요
#     pass

def solution(s: str) -> int:
    count = {} # 빈딕셔너리 생성
    for char in s: # 문자열 한글자씩 순회
        if char in count: # 이미 count에 char 가 있다면? 
            count[char] = count[char] +1 # 개수 +1 
        else:
            count[char] = 1 # 없다면 값을 1로 시작
````

---

## 테스트 케이스

````python
# 테스트 실행
def test_solution():
    # 테스트 케이스 1
    result1 = solution("leetcode")
    assert result1 == 0, f"테스트 1 실패: {result1}"
    print("테스트 1 통과!")
    
    # 테스트 케이스 2
    result2 = solution("loveleetcode")
    assert result2 == 2, f"테스트 2 실패: {result2}"
    print("테스트 2 통과!")
    
    # 테스트 케이스 3
    result3 = solution("aabb")
    assert result3 == -1, f"테스트 3 실패: {result3}"
    print("테스트 3 통과!")
    
    print("\n모든 테스트 통과!")

test_solution()
````

### 풀이 1: Counter 사용
```python
from collections import Counter

def solution(s):
    count = Counter(s)  # 각 문자 개수 세기
    
    for i, char in enumerate(s):
        if count[char] == 1:
            return i
    return -1
```

### 풀이 2: dict 직접 사용
```python
def solution(s):
    count = {}
    
    # 1단계: 개수 세기
    for char in s:
        count[char] = count.get(char, 0) + 1
    
    # 2단계: 개수가 1인 첫 번째 문자 찾기
    for i, char in enumerate(s):
        if count[char] == 1:
            return i
    return -1
```

### 핵심 포인트
- 1단계: 모든 문자 개수 세기 (HashMap)
- 2단계: 순서대로 확인하면서 개수 1인 것 찾기
- `count.get(char, 0)`: char가 없으면 0 반환
