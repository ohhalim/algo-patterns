## Part 1: 고정 길이 윈도우 (Fixed Window)

### 문제 1-1: 연속된 k개 합의 최댓값
정수 리스트 `nums`와 `k`가 주어질 때, 길이 `k`인 연속 구간의 합 중 최댓값을 반환하세요.

````python
# 함수정의
# 입력: 정수리스트 nums, 구간길이 k
# 출력: 길이 k 인연속구간 합의 최댓값 int
def max_sum_k_elements(nums: list, k: int) -> int:
    # 힌트: 첫 윈도우 합을 만든 뒤,
    # (빠지는 값 빼기 + 새로 들어오는 값 더하기)로 이동

    # 예외처리 k가 0이하이거나 배열의 길이보다 크면 유효한 구간이 없으니 0반환
    if k <= 0 or k > len(nums): 
        return 0

    # 첫 윈도우의 합을 계산
    window_sum = sum(nums[:k])
    # 현재까지 최댓값 초기화, 처음엔 윈도우밖에 못봤으니 그 값을 최댔값으로 둠
    max_sum = window_sum

    # 윈도우를 오른쪽으로 한칸씩 이동하는 반복
    # i 는 새로들어오는 원소의 인덱스 
    # 시작이 k인 이유 0~k-1 은 이미 첫 윈도우로 계산했기 떄문에
    for i in range(k, len(nums)):
        window_sum = window_sum - nums[i - k] + nums[i]
        max_sum =max(max_sum, window_sum)
    
    return max_sum


# 테스트
print(max_sum_k_elements([2, 1, 5, 1, 3, 2], 3))  # 9
print(max_sum_k_elements([1, 1, 1, 1], 2))        # 2
````

def max_average_subarray(nums: list, k: int) -> float:
    if k <= 0 or k > len(nums):
        return 0.0

    window_sum = sum(nums[:k])
    max_sum = window_sum

    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]
        max_sum = max(max_sum, window_sum)

    return round(max_sum / k, 2)

def min_subarray_len(target: int, nums: list) -> int:
    n = len(nums)
    left = 0
    curr_sum = 0
    best = n + 1

    for right in range(n):
        curr_sum += nums[right]

        while curr_sum >= target:
            best = min(best, right - left + 1)
            curr_sum -= nums[left]
            left += 1

    return 0 if best == n + 1 else best

def min_subarray_len_with_indices(target: int, nums: list) -> tuple:
    n = len(nums)
    left = 0
    curr_sum = 0
    best = n + 1
    best_left, best_right = -1, -1

    for right in range(n):
        curr_sum += nums[right]

        while curr_sum >= target:
            length = right - left + 1
            if length < best:
                best = length
                best_left = left
                best_right = right

            curr_sum -= nums[left]
            left += 1

    if best == n + 1:
        return (0, -1, -1)

    return (best, best_left, best_right)

def length_of_longest_substring(s: str) -> int:
    left = 0
    seen = set()
    best = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1

        seen.add(s[right])
        best = max(best, right - left + 1)

    return best

from collections import Counter

def find_anagram_starts(s: str, p: str) -> list:
    if len(p) > len(s):
        return []

    result = []
    p_count = Counter(p)
    window_count = Counter(s[:len(p)])

    if window_count == p_count:
        result.append(0)

    for right in range(len(p), len(s)):
        left_char = s[right - len(p)]
        right_char = s[right]

        window_count[left_char] -= 1
        if window_count[left_char] == 0:
            del window_count[left_char]

        window_count[right_char] += 1

        if window_count == p_count:
            result.append(right - len(p) + 1)

    return result
