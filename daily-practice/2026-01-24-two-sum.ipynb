## 문제: 두 수의 합 (Two Sum)

### 난이도: Easy

---

### 문제 설명

정수 배열 `nums`와 정수 `target`이 주어집니다.  
배열에서 **두 수를 더해서 `target`이 되는 두 수의 인덱스**를 반환하세요.

- 각 입력에는 **정확히 하나의 해답**만 존재합니다.
- 같은 요소를 두 번 사용할 수 없습니다.
- 답은 어떤 순서로 반환해도 됩니다.

---

### 예제

**예제 1:**
```
입력: nums = [2, 7, 11, 15], target = 9
출력: [0, 1]
설명: nums[0] + nums[1] = 2 + 7 = 9
```

**예제 2:**
```
입력: nums = [3, 2, 4], target = 6
출력: [1, 2]
설명: nums[1] + nums[2] = 2 + 4 = 6
```

**예제 3:**
```
입력: nums = [3, 3], target = 6
출력: [0, 1]
```

---

### 제약 조건

- `2 <= nums.length <= 10^4`
- `-10^9 <= nums[i] <= 10^9`
- `-10^9 <= target <= 10^9`
- 정확히 하나의 유효한 답이 존재합니다.

---

### 힌트

1. 가장 단순한 방법은? O(n^2) 이중 반복문
2. 더 효율적인 방법은? HashMap(Dictionary)을 활용해보세요!

````python
from typing import List

# 정수 리스트와 정수를 받아서, 정수 리스트를 반환한다
def two_sum(nums: List[int], target: int) -> List[int]:
    """
    두 수의 합이 target이 되는 두 수의 인덱스를 반환합니다.
    
    Args:
        nums: 정수 배열 
        target: 목표 합
    
    Returns:
        두 수의 인덱스를 담은 리스트
    """
    # 여기에 코드를 작성하세요
    # for i in range(len(nums)):
    #     for j in range(i + 1, len(nums)):
    #         if nums[i] + nums[j] == target:
    #             return [i, j]

    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i


    
````

---

## 테스트 케이스

````python
# 테스트 실행
def test_two_sum():
    # 테스트 케이스 1
    result1 = two_sum([2, 7, 11, 15], 9)
    assert sorted(result1) == [0, 1], f"테스트 1 실패: {result1}"
    print("테스트 1 통과!")
    

    
    # 테스트 케이스 2
    result2 = two_sum([3, 2, 4], 6)
    assert sorted(result2) == [1, 2], f"테스트 2 실패: {result2}"
    print("테스트 2 통과!")
    
    # 테스트 케이스 3
    result3 = two_sum([3, 3], 6)
    assert sorted(result3) == [0, 1], f"테스트 3 실패: {result3}"
    print("테스트 3 통과!")
    
    print("\n모든 테스트 통과!")

test_two_sum()
````

### 풀이 1: 브루트 포스 O(n^2)
```python
def two_sum(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
```

### 풀이 2: HashMap O(n) - 추천!
```python
def two_sum(nums, target):
    num_map = {}  # 값 -> 인덱스
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
```

### 핵심 아이디어
- `target - 현재값 = 필요한 값`
- HashMap에 이미 `필요한 값`이 있으면 정답!
- 한 번의 순회로 O(n) 시간복잡도 달성
