## Part 1: for문 + 조건문 복습

### 문제 1-1: 3으로 나누어 떨어지는 수 합계
리스트에서 3의 배수만 더하기

````python
def sum_divisible_by_3(nums: list) -> int:
    result = 0
    for i in nums:
        if i % 3 == 0:
            result += int(i)
    return result 

# 테스트
print(sum_divisible_by_3([1, 3, 5, 6, 9]))  # 18 (3+6+9)
print(sum_divisible_by_3([1, 2, 4, 5]))     # 0
````

### 문제 1-2: 음수를 0으로 변환
리스트에서 음수는 0으로 바꾼 새 리스트 반환

````python
def replace_negatives(nums: list) -> list:
    result = []
    for num in nums:
        if num < 0:
            result.append(0)
        else:
            result.append(num)

    return result
    
# 테스트
print(replace_negatives([1, -2, 3, -4, 5]))  # [1, 0, 3, 0, 5]
print(replace_negatives([-1, -2, -3]))       # [0, 0, 0]
````

### 문제 1-3: 특정 값보다 큰 원소 개수
threshold보다 큰 원소 개수 반환

````python
def count_greater_than(nums: list, threshold: int) -> int:
    result = 0
    for num in nums:
        if num > threshold:
            result += 1
    return result 

# 테스트
print(count_greater_than([1, 5, 10, 15, 20], 10))  # 2 (15, 20)
print(count_greater_than([1, 2, 3], 5))            # 0
````

### 문제 2-3: 숫자 빈도수
숫자 리스트에서 각 숫자가 몇 번 나오는지

````python
def count_numbers(nums: list) -> dict:
    result = {}
    for num in nums:
        if num in result:
            result[num] += 1
        else:
            result[num] = 1
    return result 
# 흐흐 맞췄다 기분개좋아 

# 테스트
print(count_numbers([1, 2, 2, 3, 3, 3]))  # {1: 1, 2: 2, 3: 3}
print(count_numbers([5, 5, 5, 5]))        # {5: 4}
````

## Part 3: enumerate 복습

### 문제 3-1: 짝수 인덱스 원소만
인덱스가 짝수(0, 2, 4...)인 원소만 반환

````python
def even_indices(lst: list) -> list:
    result = []
    for index, value in enumerate(lst): #인덱스, 값 순이야 enumerate는 
        if index % 2 == 0:
            result.append(value)
    return result 

# 테스트
print(even_indices([10, 20, 30, 40, 50]))  # [10, 30, 50]
print(even_indices(['a', 'b', 'c', 'd']))  # ['a', 'c']
````

## Part 4: 문자열 + 슬라이싱 복습

### 문제 4-1: 문자열 끝에서 n글자
문자열 끝에서 n글자 반환

````python
def last_n_chars(s: str, n:int) -> str:
    return s[-n:]

# 테스트
print(last_n_chars("hello", 2))   # "lo"
print(last_n_chars("python", 4))  # "thon"
````

### 문제 4-2: 모음 제거
문자열에서 모음(a, e, i, o, u) 제거

````python
def remove_vowels(s: str) -> str:
    result = ""
    vowels = "aeiou"
    for char in s:
        if char not in vowels:
            result += char 
    return result 

# 테스트
print(remove_vowels("hello"))      # "hll"
print(remove_vowels("beautiful"))  # "btfl"
````

### 문제 4-3: 단어 길이 리스트
문장을 받아서 각 단어의 길이 리스트 반환

````python
def word_lengths(sentence: str) -> list:
    # 힌트: split()으로 단어 분리
    result = []
    split_words = sentence.split()
    for word in split_words:
        # result += len(word) # += 말고 append 써야함
        result.append(len(word))
        
    return result

# 테스트
print(word_lengths("I love python"))  # [1, 4, 6]
print(word_lengths("hello world"))    # [5, 5]
````

## Part 5: while문 복습

### 문제 5-1: n까지 짝수 합
1부터 n까지 짝수만 더하기 (while문 사용)

````python
def sum_even_to_n(n: int) -> int:
    result = 0

    # while n > 0:
    #     result += n 
    #     n = n - 2
    # return result

    i = 2

    while i <= n:
        result += i
        i += 2

    return result 

# 테스트
print(sum_even_to_n(10))  # 30 (2+4+6+8+10)
print(sum_even_to_n(7))   # 12 (2+4+6)
````

### 문제 5-2: 자릿수 개수
정수의 자릿수 개수 반환

````python
def count_digits(n: int) -> int:
    # 힌트: n // 10 반복
    result = 0
    # while n >= 10:
    #     if n // 10: 
    #     result += 1
    #     n = n % 10 
    # return result 
    if n == 0: return 1 
    while n > 0:
        n = n // 10
        result += 1
    return result 


# 테스트
print(count_digits(12345))  # 5
print(count_digits(7))      # 1
````

## Part 6: 종합 문제

### 문제 6-1: 두 리스트 합집합
두 리스트의 합집합 (중복 제거)

````python
def union(list1: list, list2: list) -> list:
    return list(set(list1 | set(list2)) # | 합집합 연산자 

# 테스트
print(sorted(union([1, 2, 3], [3, 4, 5])))  # [1, 2, 3, 4, 5]
print(sorted(union([1, 1, 2], [2, 3])))     # [1, 2, 3]
````

### 문제 6-3: 공통 문자 찾기
두 문자열에 공통으로 있는 문자들 반환

````python
def common_chars(s1: str, s2: str) -> str:
    result = ''
    for char in s1:
        if char in s2 and char not in result:
            result += char
    return result
    

# 테스트
print(common_chars("hello", "world"))  # "lo" 또는 "ol"
print(common_chars("abc", "xyz"))      # ""
````
