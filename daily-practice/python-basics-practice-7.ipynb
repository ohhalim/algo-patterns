## Part 1: 2중 반복문 (Nested Loops)

### 문제 1-1: 구구단 출력
2단부터 9단까지 출력 (리스트로 반환)

````python
def multiplication_table() -> list:
    # 힌트: 바깥 for문은 단(2~9), 안쪽 for문은 곱하는 수(1~9)
    result = []
    for i in range(2, 10):
        for j in range(1, 10):
            result.append(f"{i} x {j} = {i * j}")
    return result

# 테스트
result = multiplication_table()
print(result[:3])  # ['2 x 1 = 2', '2 x 2 = 4', '2 x 3 = 6']
````

### 문제 1-2: 모든 쌍의 합
두 리스트의 모든 조합의 합 구하기

````python
def all_pair_sums(list1: list, list2: list) -> list:
    # 힌트: 2중 for문으로 모든 조합 만들기
    result = []
    for num1 in list1:
        for num2 in list2:
            result.append(num1 + num2)
    return result 

# 테스트
print(all_pair_sums([1, 2], [3, 4]))  # [4, 5, 5, 6]
print(all_pair_sums([1], [1, 2, 3]))  # [2, 3, 4]
````

### 문제 1-3: 2D 리스트 합계
2차원 리스트의 모든 원소 합

````python
def sum_2d_list(matrix: list) -> int:
    result = 0
    for row in matrix:
        for num in row:
            result += num
    return result

# 테스트
print(sum_2d_list([[1, 2], [3, 4]]))        # 10
print(sum_2d_list([[1, 2, 3], [4, 5, 6]]))  # 21
````

## Part 2: 리스트 컴프리헨션 (List Comprehension)

### 문제 2-1: 제곱 리스트
1부터 n까지 숫자의 제곱 리스트

````python
def squares(n: int) -> list:
    # 힌트: [x*x for x in range(...)]
    # 이거 뜻이 n의 범위를 1부터 n까지 갈때 각각 원소를 제곱해라
    return [x*x for x in range(n)]


# 테스트
print(squares(5))   # [1, 4, 9, 16, 25]
print(squares(3))   # [1, 4, 9]
````

### 문제 2-2: 조건부 필터링
리스트에서 5보다 큰 수만 2배로

````python
def double_if_greater(nums: list) -> list:
    # 힌트: [x*2 for x in nums if x > 5]
    # 이거 뜻이 nums을 순회하면서 x가 5보다 크면 x*2를 해서 리스트에 추가하는 거야
    return [x*2 for x in nums if x > 5]


# 테스트
print(double_if_greater([1, 6, 3, 8, 2]))  # [12, 16]
print(double_if_greater([1, 2, 3]))        # []
````

## Part 3: 재귀 기초 (Recursion Basics)

### 문제 3-1: 팩토리얼
n! = n × (n-1) × ... × 1

````python
def factorial(n: int) -> int:
    # 힌트: 기저 조건 n==1일 때 1 반환, 아니면 n * factorial(n-1)
    if n == 1:
        return 1
    return n * factorial(n-1)
    

# 테스트
print(factorial(5))  # 120 (5*4*3*2*1)
print(factorial(3))  # 6 (3*2*1)
````

### 문제 3-2: 리스트 합 (재귀로)
재귀를 사용해서 리스트의 합 구하기

````python
def recursive_sum(nums: list) -> int:
    # 힌트: 빈 리스트면 0, 아니면 첫 원소 + recursive_sum(나머지)
    if not nums:
        return 0 
    return nums[0] + recursive_sum(nums[1:])

# 테스트
print(recursive_sum([1, 2, 3, 4]))  # 10
print(recursive_sum([5]))           # 5
````

## Part 4: 투 포인터 (Two Pointers)

### 문제 4-1: 두 수의 합 찾기
정렬된 리스트에서 합이 target인 두 수의 인덱스

````python
def two_sum_sorted(nums: list, target: int) -> list:
    # 힌트: left=0, right=len(nums)-1로 시작
    # 합이 target보다 작으면 left++, 크면 right--
    left = 0
    right = len(nums) - 1

    while left < right:
        if nums[left] + nums[right] == target:
            return [left, right]
        elif nums[left] + nums[right] < target:
            left += 1
        else:
            right -= 1
    return []

# 테스트
print(two_sum_sorted([1, 2, 3, 4, 5], 7))  # [1, 4] (2+5=7)
print(two_sum_sorted([1, 3, 5, 7], 10))    # [1, 3] (3+7=10)
````

### 문제 4-2: 회문 검사 (투 포인터)
양 끝에서 시작해서 가운데로 이동하며 비교

````python
def is_palindrome_two_pointer(s: str) -> bool:
    # 힌트: left, right 포인터를 양 끝에서 시작
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

# 테스트
print(is_palindrome_two_pointer("racecar"))  # True
print(is_palindrome_two_pointer("hello"))    # False
````

## Part 5: 슬라이딩 윈도우 기초

### 문제 5-1: 연속된 k개 합의 최댓값
연속된 k개 원소의 합 중 최댓값

````python
def max_sum_k_elements(nums: list, k: int) -> int:
    # 힌트: 처음 k개 합을 구하고, 한 칸씩 윈도우 이동

    # 풀이
    window_sum = sum(nums[:k])
    max_sum = window_sum

    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]
        max_sum = max(max_sum, window_sum)

    return max_sum  

# 테스트
print(max_sum_k_elements([1, 4, 2, 10, 2, 3, 1], 3))  # 16 (4+2+10)
print(max_sum_k_elements([1, 2, 3, 4, 5], 2))         # 9 (4+5)
````

## Part 6: 해시 심화 (딕셔너리 고급)

### 문제 6-1: 두 수의 합 (해시)
리스트에서 합이 target인 두 수의 인덱스 (정렬 안 됨)

````python
def two_sum_hash(nums: list, target: int) -> list:
    # 힌트: {값: 인덱스} 딕셔너리 사용
    # target - 현재값이 딕셔너리에 있는지 확인

    result = []
    # 이거 뜻이 nums를 순회하면서 target - nums[i]가 result에 있는지 확인하고
    # 있다면 그 인덱스와 현재 인덱스를 반환하는 거야
    for i in range(len(nums)):
        if target - nums[i] in result:
            return [result.index(target - nums[i]), i]
        result.append(nums[i])
    return []   

# 테스트
print(two_sum_hash([2, 7, 11, 15], 9))   # [0, 1] (2+7=9)
print(two_sum_hash([3, 2, 4], 6))        # [1, 2] (2+4=6)
````

### 문제 6-2: 문자 재배열 (Anagram)
두 문자열이 같은 문자로 구성되어 있는지

````python
from collections import Counter

def is_anagram(s1: str, s2: str) -> bool:
    # 힌트: 각 문자열의 문자 빈도수를 딕셔너리로 세고 비교
    return Counter(s1) == Counter(s2)

# 테스트
print(is_anagram("listen", "silent"))  # True
print(is_anagram("hello", "world"))    # False
````

## Part 7: 스택 패턴 준비

### 문제 7-1: 괄호 짝 맞추기
여는 괄호와 닫는 괄호가 올바르게 짝지어졌는지

````python
def valid_parentheses(s: str) -> bool:
    # 힌트: 리스트를 스택처럼 사용 (append, pop)
    # '(' 나오면 push, ')' 나오면 pop하고 비교
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack:
                return False
            stack.pop()
    return not stack

# 테스트
print(valid_parentheses("(())"))    # True
print(valid_parentheses("(()"))     # False
print(valid_parentheses("()()"))    # True
````

### 문제 7-2: 역순 문자열 (스택 이용)
스택을 사용해서 문자열 뒤집기

````python
def reverse_with_stack(s: str) -> str:
    # 힌트: 한 글자씩 push, 그 다음 전부 pop
    stack = []
    for char in s:
        stack.append(char)
    
    reversed_str = ""
    while stack:
        reversed_str += stack.pop()
    
    return reversed_str

# 테스트
print(reverse_with_stack("hello"))  # "olleh"
print(reverse_with_stack("world"))  # "dlrow"
````
