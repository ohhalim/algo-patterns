## 문제 1: 완주하지 못한 선수 (프로그래머스 Lv.1)

> 마라톤 참가자 목록 `participant`와 완주자 목록 `completion`이 주어진다.
> 완주하지 못한 **단 한 명의 선수** 이름을 반환하라. (동명이인 있을 수 있음)
>
> - `participant=["leo", "kiki", "eden"], completion=["eden", "kiki"]` → `"leo"`
> - `participant=["marina", "josipa", "nikola", "vinko", "filipa"], completion=["josipa", "filipa", "marina", "nikola"]` → `"vinko"`

**핵심 아이디어**: 참가자와 완주자를 각각 Counter로 세고, 차이를 구하면 된다.

```python
from collections import Counter

def find_non_finisher(participant: list[str], completion: list[str]) -> str:
    # 참가자 이름별로 인원스를 저장할 빈 딕셔너리를 생성합니다
    # 여기에 코드를 작성하세요

    # participant 리스트에 잇는 모든 참가자 이름을 하나씩 꺼내어 반복합니다
    count = {}
    for name in participant:
        # count 딕셔너리에 name에 해당하는 값을 가져옵니다
        # 만약 name 이 없다면 기본값 0을 가져온디 뒤에 1 을 더해 다시 저장
        # 결과적으로 모든 참가자의 이름 을 키로, 사람수를 값으로 저장  
        count[name] = count.get(name, 0) + 1
    # comletion 리스트에 있는 완주자이름을 하나씩 꺼내어 반복합니다
    for name in completion:
        # 완주한 사람의 이름이 들어오면, 아까 카운트 해둔 딕셔너리에 해당 이름의 사람 수를 1명 씩 뺍니다
        count[name] -= 1
    # count 딕셔너리에 있는 모든키와 값의 쌍을 반복해서 꺼냅니다
    for name, cnt in count.item():
        # 만약 값이 0보다 크다면 완주하지 못한 사람이니 
        if cnt > 0:
            # 완주하니 못한 사람을 출력하고 함수 종료
            return name

    # count = {}
    # for name in participant:
    #     count[name] = count.get(name, 0) + 1
    # for name in completion:
    #     count[name] -= 1
    # for namem cnt in count,item():
    #     if cnt > 0:
    #         return name
def solution(participant: list[str], completion: list[str]) -> str:
    remain = Counter(participant) - Counter(completion)
    return next(iter(remain))
    
# 테스트
assert find_non_finisher(["leo", "kiki", "eden"], ["eden", "kiki"]) == "leo"
assert find_non_finisher(["marina", "josipa", "nikola", "vinko", "filipa"], ["josipa", "filipa", "marina", "nikola"]) == "vinko"
assert find_non_finisher(["mislav", "stanko", "mislav", "ana"], ["stanko", "ana", "mislav"]) == "mislav"
print("모든 테스트 통과!")
```
