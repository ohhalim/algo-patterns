## 문제 1: 완주하지 못한 선수 (프로그래머스 Lv.1)

> 마라톤 참가자 목록 `participant`와 완주자 목록 `completion`이 주어진다.
> 완주하지 못한 **단 한 명의 선수** 이름을 반환하라. (동명이인 있을 수 있음)
>
> - `participant=["leo", "kiki", "eden"], completion=["eden", "kiki"]` → `"leo"`
> - `participant=["marina", "josipa", "nikola", "vinko", "filipa"], completion=["josipa", "filipa", "marina", "nikola"]` → `"vinko"`

**핵심 아이디어**: 참가자와 완주자를 각각 Counter로 세고, 차이를 구하면 된다.

```python
from collections import Counter

def find_non_finisher(participant: list[str], completion: list[str]) -> str:
    # 참가자 이름별로 인원스를 저장할 빈 딕셔너리를 생성합니다
    # 여기에 코드를 작성하세요

    # participant 리스트에 잇는 모든 참가자 이름을 하나씩 꺼내어 반복합니다
    count = {}
    for name in participant:
        # count 딕셔너리에 name에 해당하는 값을 가져옵니다
        # 만약 name 이 없다면 기본값 0을 가져온디 뒤에 1 을 더해 다시 저장
        # 결과적으로 모든 참가자의 이름 을 키로, 사람수를 값으로 저장  
        count[name] = count.get(name, 0) + 1
    # comletion 리스트에 있는 완주자이름을 하나씩 꺼내어 반복합니다
    for name in completion:
        # 완주한 사람의 이름이 들어오면, 아까 카운트 해둔 딕셔너리에 해당 이름의 사람 수를 1명 씩 뺍니다
        count[name] -= 1
    # count 딕셔너리에 있는 모든키와 값의 쌍을 반복해서 꺼냅니다
    for name, cnt in count.item():
        # 만약 값이 0보다 크다면 완주하지 못한 사람이니 
        if cnt > 0:
            # 완주하니 못한 사람을 출력하고 함수 종료
            return name

    # count = {}
    # for name in participant:
    #     count[name] = count.get(name, 0) + 1
    # for name in completion:
    #     count[name] -= 1
    # for namem cnt in count,item():
    #     if cnt > 0:
    #         return name
def solution(participant: list[str], completion: list[str]) -> str:
    remain = Counter(participant) - Counter(completion)
    return next(iter(remain))
    
# 테스트
assert find_non_finisher(["leo", "kiki", "eden"], ["eden", "kiki"]) == "leo"
assert find_non_finisher(["marina", "josipa", "nikola", "vinko", "filipa"], ["josipa", "filipa", "marina", "nikola"]) == "vinko"
assert find_non_finisher(["mislav", "stanko", "mislav", "ana"], ["stanko", "ana", "mislav"]) == "mislav"
print("모든 테스트 통과!")
```

## 문제 2: 위장 (프로그래머스 Lv.2)

> 스파이가 다양한 종류의 옷을 가지고 있다. `clothes = [[의상이름, 종류], ...]` 가 주어질 때,
> 서로 다른 옷의 조합 **수**를 반환하라. (하루에 옷을 반드시 하나 이상 착용)
>
> - `clothes = [["yellow_hat", "headgear"], ["blue_sunglasses", "eyewear"], ["green_turban", "headgear"]]` → `5`
>   - headgear 2개, eyewear 1개 → (2+1) * (1+1) - 1 = 5

**핵심 아이디어**: 종류별 개수를 센 뒤, `(각 종류 수 + 1)`을 모두 곱하고 1(아무것도 안 입는 경우)을 빼면 된다.

```python
from collections import defaultdict

def disguise(clothes: list[list[str]]) -> int:
    # 여기에 코드를 작성하세요
    # 옷의 종류별 개수를 저장할 빈딕셔너리를 만든다
    count = {}
    # 제공된 옷 리스트를 순회 하며 각 종류별 옷이 몇개인지 카운트
    # clothes 안에는 ["모자, "얼굴"] 처럼 이름과 종류가 들어있음
    for name, category in clothes:
        # 기존에 있넌 종류면 +1, 처음보는 종류면 0에서 시작해서 +1을 합니다
        count[category] = count.get(category, 0) +1
    # 경우의 수를 계산할 준비를 한다 
    result = 1
    # 딕셔너리에 저장된 값들만 꺼내서 순회
    for cnt in count.values():
        # 온을 안입은경우를 포함에서 모두 곱해준다
        result *=(cnt+1)
        # 모든옷을 하나도 입지 않은 예외적인 경우 (1가지)를 빼고 반환합니다
    return result -1 

# def solution(clothes: list[list[str]]) -> int:
#     count = Counter(categort for name, category in clothes)
#     result = 1 
#     for cnt in count.values():
#         result *=(cnt +1)
#     return result -1 


# 테스트
assert disguise([["yellow_hat", "headgear"], ["blue_sunglasses", "eyewear"], ["green_turban", "headgear"]]) == 5
assert disguise([["crow_mask", "face"], ["blue_sunglasses", "face"], ["smoky_makeup", "face"]]) == 3
assert disguise([["yellowhat", "headgear"]]) == 1
print("모든 테스트 통과!")
```

## 문제 3: 전화번호 목록 (프로그래머스 Lv.2)

> 전화번호 목록이 주어질 때, 어떤 번호가 다른 번호의 **접두어인지** 확인하라.
> 접두어인 번호가 하나라도 있으면 `False`, 없으면 `True`를 반환하라.
>
> - `["119", "97674223", "1195524421"]` → `False` (119는 1195524421의 접두어)
> - `["123", "456", "789"]` → `True`

**핵심 아이디어**: 모든 번호를 Set에 넣고, 각 번호의 모든 접두어가 Set에 있는지 확인한다.

```python
def phone_book(phone_list: list[str]) -> bool:
    # 여기에 코드를 작성하세요
    # 모든 전화번호를 해시를 사용하는 Set구조로 변환합니다
    phon_set = set(phone_list)
    # 전화번호부의 번호를 하나씩 꺼냅니다
    for pthone in phone_list:
        # 그번호의 길이만큽 반보하면서 번호를앞에서부터 1글자 2글자씩 잘라봅니다
        # 인덱스1 부터 시작하므로 처음엔 1 다음엔 11 다음엔 119 현테가 됩니다
        for i in range(1, len(phone)):
            # 방븝 자른 문자열이 우리가 처음에 만든 해시안에 들어있는지 확인합니다
            if phone[:i] in phone_set:
                # 발견되었다면 누군가의 번호가 내번호의 접두어라는뜻이 므로 즉히 False를 반환합ㄴ디ㅏ
                return False
    # 모든 번호의 접두어를 다잘라서 검사했는데도 걸리는게 없다면 모든 아전하무로 True를 반환합니다
    return True

# 테스트
assert phone_book(["119", "97674223", "1195524421"]) == False
assert phone_book(["123", "456", "789"]) == True
assert phone_book(["12", "123", "1235", "567", "88"]) == False
print("모든 테스트 통과!")
```
