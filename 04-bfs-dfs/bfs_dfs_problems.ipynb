# 04. BFS / DFS

**핵심**: 탐색. "전부 방문해야 하면" DFS, "최단이면" BFS

## 언제 BFS/DFS를 쓰냐
- 격자(미로, 섬) 탐색 → BFS/DFS
- 최단 거리/최소 횟수 → BFS
- 모든 경로/경우의 수 → DFS (백트래킹)
- 연결 요소 개수 → DFS/BFS

## 템플릿
```python
# BFS (최단거리)
from collections import deque
def bfs(grid, start):
    queue = deque([start])
    visited = {start}
    while queue:
        x, y = queue.popleft()
        for dx, dy in [(0,1),(0,-1),(1,0),(-1,0)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < len(grid) and 0 <= ny < len(grid[0]):
                if (nx, ny) not in visited and grid[nx][ny] == 1:
                    visited.add((nx, ny))
                    queue.append((nx, ny))

# DFS (재귀)
def dfs(grid, x, y, visited):
    if x < 0 or x >= len(grid) or y < 0 or y >= len(grid[0]):
        return
    if (x, y) in visited or grid[x][y] == 0:
        return
    visited.add((x, y))
    for dx, dy in [(0,1),(0,-1),(1,0),(-1,0)]:
        dfs(grid, x + dx, y + dy, visited)
```

# 트리(Tree)란?

## 트리의 핵심 개념

트리는 **위에서 아래로 뻗어나가는 계층 구조**입니다. 폴더 안에 폴더가 있고, 그 안에 또 폴더가 있는 컴퓨터의 파일 시스템과 같습니다.

```
        1          <- 루트(root): 맨 꼭대기 노드. 부모가 없는 유일한 노드.
       / \
      2   3        <- 자식(children): 1의 자식은 2와 3.
     / \   \
    4   5   6      <- 리프(leaf): 자식이 없는 맨 끝 노드. 4, 5, 6이 리프.
```

### 용어 정리
| 용어 | 설명 |
|------|------|
| **노드(Node)** | 트리를 구성하는 각 데이터 요소 (위 그림에서 1~6 각각) |
| **루트(Root)** | 맨 위에 있는 시작 노드 (위 그림에서 1) |
| **부모(Parent)** | 어떤 노드 바로 위에 연결된 노드 (2의 부모는 1) |
| **자식(Children)** | 어떤 노드 바로 아래에 연결된 노드들 (1의 자식은 2, 3) |
| **리프(Leaf)** | 자식이 하나도 없는 맨 끝 노드 (4, 5, 6) |
| **깊이(Depth)** | 루트에서 해당 노드까지 내려가는 간선의 수 (루트의 깊이 = 0) |
| **높이(Height)** | 트리의 최대 깊이 (위 트리의 높이 = 2) |

### 이진 트리 (Binary Tree)
코딩 테스트에서 가장 많이 나오는 트리 형태입니다.  
규칙은 단 하나: **각 노드의 자식이 최대 2개** (왼쪽 자식, 오른쪽 자식)

### BFS/DFS와 트리의 관계
앞에서 풀었던 격자(grid) 문제들은 상하좌우 4방향을 탐색했습니다.  
트리에서는 **왼쪽 자식, 오른쪽 자식** 2방향을 탐색합니다.  
탐색 원리는 완전히 동일합니다!

| | 격자(Grid) | 트리(Tree) |
|---|---|---|
| **탐색 방향** | 상, 하, 좌, 우 (4방향) | 왼쪽 자식, 오른쪽 자식 (2방향) |
| **방문 체크** | visited set 필요 | 부모->자식 한 방향이라 불필요 |
| **DFS** | 깊이 우선으로 한 경로 끝까지 | 재귀로 왼쪽 끝까지 갔다가 돌아옴 |
| **BFS** | 같은 거리의 칸을 먼저 | 같은 레벨(깊이)의 노드를 먼저 |

### 트리 DFS 템플릿 (재귀)
```python
def dfs(node):
    if node is None:     # 바닥(빈 노드)에 도달하면 멈춤
        return
    # node.val 로 현재 노드의 값 사용
    dfs(node.left)       # 왼쪽 자식으로 내려감
    dfs(node.right)      # 오른쪽 자식으로 내려감
```

---

````python
# 이진 트리 노드 정의 (트리 문제에서 공통으로 사용)
# LeetCode에서 기본 제공하는 구조와 동일
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        # val: 이 노드가 담고 있는 값 (숫자)
        self.val = val
        # left: 왼쪽 자식 노드 (없으면 None)
        self.left = left
        # right: 오른쪽 자식 노드 (없으면 None)
        self.right = right

# 테스트용 헬퍼: 리스트 -> 트리 변환
from collections import deque
def build_tree(values):
    if not values or values[0] is None:
        return None
    root = TreeNode(values[0])
    queue = deque([root])
    i = 1
    while queue and i < len(values):
        node = queue.popleft()
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root

# 테스트용 헬퍼: 트리 -> 리스트 변환
def tree_to_list(root):
    if not root:
        return []
    result = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return result

print('TreeNode 및 헬퍼 함수 준비 완료!')
````
