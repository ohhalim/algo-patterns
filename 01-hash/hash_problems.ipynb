## 2. Valid Anagram (LC 242) - Easy
https://leetcode.com/problems/valid-anagram/

**패턴**: 빈도수 비교 (Counter)

**문제**: 두 문자열 s, t가 주어질 때, t가 s의 아나그램인지 판별하라.

```
Input: s = "anagram", t = "nagaram" → True
Input: s = "rat", t = "car" → False
```

````python
def is_anagram(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
    seen = {}

    for char in s: 
        if char in seen:
            seen[char] += 1
        else: 
            seen[char] = 1
    
    for char in t:
        if char in seen:
            seen[char] -= 1
            if seen[char] < 0:
                return False
        else:
            return False

    for count in seen.values():
        if count != 0:
            return False
            
    return True
````

````python
def is_anagram(s: str, t: str) -> bool:
    # 만약 문자열 s의 총길이와 문자열 t의 총길이가 서로 같지 않다면 :
    if len(s) != len(t):
        # False 반환하고 종료 
        return False

    # 빈 딕셔너리를 만들어 seen으로 이름표 붙인다
    seen = {}


    # 첫번쩨 문잘열 s 에서 글자 char를  하나씩 차례대로 꺼내오면서 반복 한다 
    for char in s: 
        # 만약 방금 꺼낸 글자가 딕서너리의 안에 이미 들어와있다면
        if char in seen:
            # 해당 글자의 카운트를 하나 올려준다
            seen[char] += 1
        # 만약 이 글자가 딕셔너리 생전 처음 등장하는 것이라면 
        else:
            # 딕셔너리에 새롭게 공간을 파주고 개수를 1개로 설정한다
            seen[char] = 1
        
    # 이제 두번째 문자열 t 에서 글자 char 를 하나씩 차례대로 꺼내오면서 반복한다
    for char in t:
        # 만약 방금 꺼낸 t 의 글자가, 우리가 만들어 둔 딕셔너리 (seen) 에 존재한다면 
        if char in seen:
            # 재료를 하나 썼다는 뜻이니까, 기록해둔 개수에서 1개를 차감한다
            seen[char] -= 1
        # 만약 개수를 차감한 직후 확인해보니 남은 개수가 0 밑으로 떨어졌다면
        # 즉 s에 있던 개수보다 t에서 더 많이 꺼내 쓰려고 하는 상황이라면
        if seen[char] < 0:
            # 짝이 맞지 않으므로 즉시 False 반환
            return False
        # 만약 방금 꺼낸 t의 글자가, 애초에 s에는 존재하지 않던 글자라면 
        else:
            # 즉시 False 반환
            return False
    # 위의 모든 반복믄을 걸리지 않고 통과했다면 t, s는 같으니까 True 반환하고 종료
    return True
    

````
