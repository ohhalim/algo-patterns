## 3. Group Anagrams (LC 49) - Medium
https://leetcode.com/problems/group-anagrams/

**패턴**: sorted 키로 dict 그룹핑

**문제**: 문자열 배열이 주어질 때, 아나그램끼리 그룹으로 묶어 반환하라.

```
Input: strs = ["eat","tea","tan","ate","nat","bat"]
Output: [["bat"],["nat","tan"],["ate","eat","tea"]]
```



````python
from typing import List

def group_anagrams(strs: List[str]) -> List[List[str]]:
    # 문자열 묶음을 저장할 해시맵을 생성한다
    groups = {}
    # 입력으로 받은 문자열 배열을 하나씩 꺼내어 반복
    for s in strs:
        # 현제 문자열(s)의 알파벳을 정렬(sorted)한뒤 다시 하나의 문자열로 합쳐서 "".join 키(Key)로 만듭니다
        key = "".join(sorted(s))
        # 만약 이 키(key)가 딕셔너리 안에 아직 존재하지 않는다면
        if key not in groups:
            # 해당 키에 해당하는 값을 빈리스트로 초기화 한다
            groups[key] = []
        
        # 행당 키의 리스트 안에 원래 문자열 s를 추가한다
        groups[key].append(s)
    # 딕셔너리의 값인 리스트만 모아서 리스트 형태로 변환한다
    return list(groups.values())
````

## 5. Longest Consecutive Sequence (LC 128) - Medium
https://leetcode.com/problems/longest-consecutive-sequence/

**패턴**: set으로 시작점 판별

**문제**: 정렬되지 않은 정수 배열에서 가장 긴 연속 수열의 길이를 반환하라. O(n)에 풀어야 한다.

```
Input: nums = [100, 4, 200, 1, 3, 2]
Output: 4  # [1, 2, 3, 4]
```

````python
from typing import List

def longest_consecutive(nums: List[int]) -> int:
    # 배열을 집합으로 변환하여 중복을 제거하고 빠른 조회가 가능하도록 한다
    num_set = set(nums)
    # 가장 긴 수열의 길이를 저장할 변수를 0으로 초기화한다
    longest = 0
    # 집합 의 요소들을 하나씩 차례대러 꺼내며 반복한다
    for num in num_set:
        # 만약 현재 숫자에서 1을 뺀 값(num -1)이 집합에 존재하지 않는다면 (즉 이 숫자가 연속된 수열의 첫 시작점이라면)
        if (num - 1) not in num_set:
            # 현재 확인 중인 수열의 길이를 1 로 설정한다
            length = 1
            # 현재 숫자(num) 더하기 길이(length)를 한 값이 집합에 존재하는 동안 계속 반복한다
            while (num + length) in num_set:
                # 다음 연속된 수자가 존재하므로 길이를 1 증가시킨다
                length += 1
            #지금까지 기록된 가장 긴 길이와 방금 계산한 길이 중 더 큰값으로 갱신한다
            lengest = max(lengest, length)
    return longest
````
